# Preprocesamiento de texto

El preprocesamiento de texto en NLP es la etapa en la que un texto crudo se transforma en una versión más limpia y estructurada para que pueda ser analizada por un modelo. Este proceso suele incluir la normalización (como pasar todo a minúsculas, eliminar espacios extra o caracteres especiales), la limpieza de ruido (URLs, menciones, correos, hashtags, emojis), la tokenización en palabras u oraciones, la eliminación de stopwords, y la reducción de palabras a su forma base mediante lematización o stemming. También puede implicar el uso de expresiones regulares (regex) para detectar y reemplazar patrones como números, correos electrónicos o símbolos repetidos. Todo esto permite que el texto sea más uniforme y represente mejor la información antes de ser convertido en vectores o embeddings para el análisis.

NLTK (Natural Language Toolkit) es una librería de Python diseñada para trabajar con lenguaje natural, muy utilizada en la enseñanza y en proyectos académicos. Sirve para realizar tareas como tokenización, lematización, stemming, etiquetado gramatical, análisis de concordancias y manejo de corpus de texto. Además, incluye recursos lingüísticos listos para usar, como diccionarios, stopwords y datasets, lo que la hace ideal para experimentar y aprender los fundamentos del procesamiento de texto antes de pasar a librerías más rápidas o pensadas para producción como spaCy.

### Tokenizar

Tokenizar es el proceso de dividir un texto en unidades más pequeñas llamadas tokens.
Un token puede ser una palabra, un signo de puntuación, un número o incluso una oración, dependiendo de cómo se defina la segmentación.

```python
"El NLP es increíble."
```

```python
["El", "NLP", "es", "increíble", "."]
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Descargar recursos necesarios
# nltk.download("punkt")

texto = "El NLP es increíble. Me gusta aprender Python."

# Tokenización por palabras

tokens_palabras = word_tokenize(texto)
print("Tokens por palabras:", tokens_palabras)

# Tokenización por oraciones
tokens_oraciones = sent_tokenize(texto)
print("Tokens por oraciones:", tokens_oraciones)
```

```python
import spacy

# Cargar modelo en español
nlp = spacy.load("es_core_news_sm")

texto = "El NLP es increíble. Me gusta aprender Python."

# Procesar texto
doc = nlp(texto)

# Tokenización por palabras
tokens_palabras = [token.text for token in doc]
print("Tokens por palabras:", tokens_palabras)

# Tokenización por oraciones
tokens_oraciones = [sent.text for sent in doc.sents]
print("Tokens por oraciones:", tokens_oraciones)
```

#### Tokenizar una pag web

```python
from bs4 import BeautifulSoup
 
import urllib.request
```

```python
url = "https://es.wikipedia.org/wiki/Finanzas"

# Agregar headers para simular un navegador
headers = {"User-Agent": "Mozilla/5.0"}

req = urllib.request.Request(url, headers=headers)

response = urllib.request.urlopen(req)  
# Abre la URL con los headers y obtiene la respuesta del servidor (Wikipedia sobre Finanzas)

html = response.read()  
# Lee el contenido HTML completo de la página web descargada

soup = BeautifulSoup(html, "html.parser")  
# Convierte el HTML en un objeto BeautifulSoup, lo que facilita analizarlo y manipularlo
# (ejemplo: quitar etiquetas, extraer párrafos, títulos, etc.)

text = soup.get_text(strip=True)  
# Extrae el texto limpio de la página, eliminando todas las etiquetas HTML
# El parámetro "strip=True" elimina espacios y saltos de línea innecesarios

print(text)  
# Muestra en consola solo los primeros 1000 caracteres del texto plano
```

```python
# Crear nuestros tokens a partir de un texto

# Tokenizamos el texto: divide el texto en palabras y signos de puntuación
tokens = word_tokenize(text)  

# Filtramos los tokens:
# 1. Convertimos cada palabra a minúsculas (para uniformidad).
# 2. Usamos isalpha() para quedarnos solo con tokens que contengan letras (así eliminamos números y signos de puntuación).
tokens = [word.lower() for word in tokens if word.isalpha()]  

# Imprimimos la lista final de tokens limpios
print(tokens)  
```

```python
nltk.FreqDist(tokens)
```

```python
# Crear la distribución de frecuencias
freq = nltk.FreqDist(tokens)

# Mostrar palabras ordenadas de mayor a menor frecuencia
for key, val in freq.most_common():
    print(str(key) + ": " + str(val))
```

```python
# Visualizar las palabras más frecuentes usando matplotlib y NLTK
import matplotlib.pyplot as plt
import nltk

# freq ya contiene la distribución de frecuencias de los tokens
# freq = nltk.FreqDist(tokens)

# Configurar el tamaño de la figura y el estilo
plt.figure(figsize=(12,6))  # ancho y alto del gráfico
plt.title("Top 30 palabras más frecuentes", fontsize=16)
plt.xlabel("Palabras", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)

# Graficar las 30 palabras más frecuentes
freq.plot(30, cumulative=False)

# Mostrar el gráfico
plt.show()
```

### Palabras de parada

Las palabras de parada o stopwords son aquellas palabras muy comunes en un idioma que, por sí solas, aportan poco significado semántico al análisis de texto. Ejemplos en español son: “el”, “la”, “de”, “y”, “que”, “en”, etc.

### Por qué son importantes?

* Aparecen con mucha frecuencia, lo que puede enmascarar patrones útiles si no se manejan adecuadamente.

* Su presencia puede inflar estadísticas de frecuencia y desviar el análisis hacia palabras que realmente no aportan información relevante.  

### Cuándo conviene quitarlas  

* En tareas como análisis de frecuencia, clasificación de texto, vectorización (BoW o TF-IDF) o análisis de sentimiento, donde queremos resaltar palabras que aportan significado al texto, suele ser conveniente eliminarlas.

### No conviene quitarlas cuando:
* Se hace análisis gramatical 

* Se necesita conservar contexto completo, por ejemplo en tareas de generación de texto o modelos basados en transformers, ya que algunas stopwords ayudan a mantener la estructura de las oraciones.

En Python, NLTK y spaCy tienen listas predefinidas de stopwords, y también se pueden crear listas personalizadas según el dominio del texto (por ejemplo, eliminar palabras muy frecuentes en correos corporativos).

### Derivación Regresiva

En muchos contextos de NLP o lingüística computacional se refiere a la técnica de reducir una palabra a su raíz o forma base eliminando sufijos o terminaciones. Esto se conoce más comúnmente como stemming.  

Por ejemplo, en español:  

“corriendo”, “corrí”, “correrán” → raíz: “corr”  

“jugaré”, “jugando”, “jugador” → raíz: “jug”

#### Para qué se usa

* Se utiliza principalmente en procesamiento de texto y recuperación de información:

* Búsqueda de información: permite que al buscar “correr”, también se encuentren documentos que contienen “corriendo” o “corrí”.

* Clasificación de texto y análisis de frecuencia: agrupa palabras con la misma raíz para evitar contar variantes como palabras distintas.

* Reducción de dimensionalidad en NLP: disminuye el tamaño del vocabulario en modelos de Bag-of-Words o TF-IDF.

#### Cuándo es útil

* Cuando se busca simplificar el texto sin perder el sentido básico de las palabras.

* Para textos con muchas variantes de una misma palabra, como blogs, noticias o redes sociales.

* No es tan útil cuando se necesita precisión gramatical, como análisis sintáctico o generación de texto, porque el stemming puede eliminar sufijos importantes y producir raíces no reconocibles (“corr” en lugar de “correr”).

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer 
```

###### PorterStemmer

* Fue diseñado originalmente para inglés.

* Es simple y rápido, pero a veces demasiado agresivo en la reducción de palabras.

```python
# Crear un objeto Stemmer usando el algoritmo de Porter
stemmer = PorterStemmer()

# Aplicar stemming a diferentes formas de la palabra "play"
print(stemmer.stem('play'))     # Salida: 'play'
print(stemmer.stem('played'))   # Salida: 'play'
print(stemmer.stem('playing'))  # Salida: 'play'
print(stemmer.stem('player'))  # Salida: 'play'
```

##### SnowballStemmer

* Es más moderno y soporta varios idiomas (incluido español).

* Menos agresivo y más preciso que Porter en muchos casos.

```python
from nltk.stem.snowball import SnowballStemmer

# Crear un stemmer para español
spanish_stemmer = SnowballStemmer('spanish')

# Palabras para aplicar stemming
words = ["trabajando", "trabajo", "trabajaría", "trabajadores"]

# Aplicar stemming
for word in words:
    print(f"{word} -> {spanish_stemmer.stem(word)}")
```

```python
# Crear un stemmer para inglés
english_stemmer = SnowballStemmer('english')

# Palabras para aplicar stemming
words = ["playing", "played", "plays", "player"]

# Aplicar stemming
for word in words:
    print(f"{word} -> {english_stemmer.stem(word)}")
```

### Palabras lematizadoras

La lematización es el proceso de reducir una palabra a su forma base o lema, que es una palabra real y reconocible en el idioma.  
  
A diferencia del stemming, que solo corta sufijos de manera aproximada, la lematización produce palabras completas correctas.

Ejemplo en español:  

“corriendo” → “correr”  

“trabajaría” → “trabajar”  

“mejores” → “mejor”  

En inglés:  

“running” → “run”  

“better” → “good”  

##### Para qué sirve  

La lematización se usa para:  

* Normalizar el texto: todas las formas de una palabra se convierten en su lema, evitando duplicados en el análisis.

* Mejorar modelos de NLP: reduce el tamaño del vocabulario en modelos Bag-of-Words, TF-IDF o embeddings.

* Análisis semántico más preciso: mantiene la forma real de la palabra, lo que permite interpretar correctamente los resultados.

* Búsqueda y recuperación de información: encontrar documentos que contengan diferentes formas de una misma palabra.

##### Cuándo se usa

* Cuando se necesita precisión semántica y queremos que la raíz sea una palabra válida del idioma.

* Clasificación de texto

* Análisis de sentimientos

* Resúmenes automáticos

* Chatbots y asistentes virtuales

* No es estrictamente necesario en tareas muy rápidas de conteo de frecuencia o clustering, donde stemming puede ser suficiente y más rápido.

##### Resumen práctico:

* Stemming → rápido, raíces aproximadas, puede producir palabras no reales.

* Lematización → más preciso, palabras reales, más costoso computacionalmente.

```python
from nltk.stem import WordNetLemmatizer
```

```python
# Crear un objeto lematizador de NLTK
lemmatizer = WordNetLemmatizer()

# Lematizar la palabra 'working' como verbo
# Por defecto, lemmatize() asume que la palabra es un sustantivo
# Al pasar pos="v" indicamos que es un verbo, así devuelve su forma base correcta

print(lemmatizer.lemmatize('working', pos="v"))  # Salida: 'work'
```

# Word embedings

Un word embedding es una representación numérica de las palabras en un espacio vectorial. Cada palabra se convierte en un vector de números, de manera que palabras con significados similares quedan cercanas entre sí en ese espacio. Por ejemplo, "king" y "queen" estarán más cerca que "king" y "banana". Esto permite que los algoritmos puedan “entender” relaciones semánticas y sintácticas entre palabras.

Los embeddings se utilizan en casi todas las tareas modernas de procesamiento de lenguaje natural. Sirven para clasificación de texto (por ejemplo, detectar spam o analizar sentimientos), búsqueda semántica de documentos relacionados, traducción automática, construcción de chatbots, y también para visualización o clustering de palabras según su significado. Básicamente, permiten que los modelos trabajen con significado y contexto de palabras, no solo con su forma.

Se usan siempre que queremos que un modelo de NLP comprenda el significado y contexto de las palabras en lugar de tratarlas como simples símbolos. Son especialmente útiles cuando se trabaja con grandes corpus de texto o cuando la relación semántica entre palabras es importante para la tarea, como en modelos de clasificación, generación de texto, traducción automática o sistemas de recomendación de contenido textual.

#### Como se hacen?  

Los word embeddings pueden obtenerse de dos formas: usando modelos pre-entrenados o entrenándolos en tu propio corpus. Los pre-entrenados, como GloVe, Word2Vec, fastText o los modelos medianos de spaCy (en_core_web_md, es_core_news_md), ya vienen con vectores listos para muchas palabras comunes. Si tu corpus es muy específico, puedes entrenar tus propios embeddings usando librerías como gensim, de manera que los vectores reflejen mejor el significado de tu vocabulario particular.

## Embeddings pre entrenados

Un embedding preentrenado es un conjunto de vectores de palabras ya entrenados en un corpus grande y diverso, como Wikipedia, noticias o Common Crawl.  

Cada palabra tiene asociado un vector numérico que captura su significado y relaciones con otras palabras.  

Lo “preentrenado” significa que otra persona o equipo ya entrenó el modelo usando técnicas como Word2Vec, GloVe o FastText, por lo que no necesitas entrenarlo tú mismo.  

#### Por qué se usa?

* Ahorras tiempo y recursos: no necesitas millones de frases ni potencia de cómputo.

* Obtienes vectores de alta calidad para palabras comunes y muchas relaciones semánticas.

* Ideal cuando tu corpus es pequeño o no tiene suficiente contexto para entrenar un Word2Vec desde cero.

La diferencia principal entre Skip-gram y CBOW radica en la dirección de predicción durante el entrenamiento de Word2Vec. Skip-gram toma una palabra central (target) y trata de predecir las palabras que aparecen a su alrededor (contexto), mientras que CBOW hace lo contrario: toma las palabras del contexto y predice la palabra central. Skip-gram es más útil cuando se trabaja con textos pequeños o con palabras poco frecuentes, ya que puede capturar relaciones semánticas finas incluso con pocas apariciones. Por otro lado, CBOW suele ser más rápido y eficiente en textos grandes, aprendiendo de manera más estable la representación de palabras frecuentes. En general, se usa Skip-gram cuando se quiere priorizar la calidad de embeddings para vocabularios raros o específicos, y CBOW cuando se necesita velocidad y se trabaja con textos extensos y vocabularios más comunes.

polarity_scores(text) devuelve un diccionario con cuatro valores:    
    

* neg:Proporción de sentimiento negativo en el texto (0 a 1)
* neu	:Proporción neutral (0 a 1)
* pos	:Proporción positiva (0 a 1)
* compound	:Score general de sentimiento (-1 muy negativo a 1 muy positivo)

Interpretar los resultados  

* compound >= 0.05 → sentimiento positivo

* compound <= -0.05 → sentimiento negativo

* -0.05 < compound < 0.05 → sentimiento neutral

```python
# Crear el analizador una sola vez
sia = SentimentIntensityAnalyzer()

def sentimiento(texto):
    # Obtener los scores del texto
    scores = sia.polarity_scores(texto)
    compound = scores['compound']  # score general entre -1 y 1

    # Clasificar según el compound
    if compound <= -0.05:
        print("El sentimiento es malo")
    elif compound >= 0.05:
        print("El sentimiento es bueno")
    else:
        print("Sentimiento neutro")
```

### N- Grams Enseñando a nuestra computadora a escribir

La técnica de n-grams consiste en identificar sub-secuencias dentro de una secuencia de datos. Esto es útil en áreas como procesamiento de lenguaje natural, secuencias de genes o secuencias de aminoácidos. Dependiendo del objetivo, los n-grams pueden construirse sobre fonemas, sílabas, letras o palabras.

Un modelo de n-grams es un modelo probabilístico que permite predecir el siguiente elemento de una secuencia a partir de los elementos anteriores. Se basa en cadenas de Markov de orden n-1, considerando que cada elemento depende de los n-1 anteriores.

Entre sus ventajas destacan su simplicidad y la facilidad para ampliar el contexto aumentando el valor de n. La idea surge de los experimentos de Claude Shannon, quien analizaba cómo predecir la siguiente letra en una secuencia de texto usando distribuciones de probabilidad.

Los n-grams también se utilizan en grandes corpus de texto, como el Google Ngram Viewer, que analiza más de cinco millones de libros de los últimos 500 años. Estos datos permiten estudiar tendencias lingüísticas, frecuencia de palabras y mejorar sistemas de recomendación de consultas.

Finalmente, los n-grams reciben nombres según su tamaño:

1-grama: unigrama

2-grama: bigrama o digrama

3-grama: trigramas

En procesamiento de lenguaje natural no basta con entender el significado de las palabras; también es crucial comprender el contexto en el que aparecen.

Los n-gramas permiten capturar este contexto, analizando secuencias de palabras o elementos para interpretar mejor el significado de lo que se escribe o se dice.

# 1- Gram  

Primero hacemos el análisis de 1-grams (unigramas) porque nos permite estudiar la frecuencia de cada palabra individual en el texto. Esto es el paso más básico para entender la distribución de palabras antes de pasar a combinaciones más complejas (como bigramas o trigramas).    
  
Cuando escribimos:  
$$P(w_i) = \frac{\text{count}(w_i)}{N}$$

significa que la probabilidad de que aparezca la palabra Wi se calcula como el número de veces que aparece esa palabra en el texto (count(w_i)) dividido entre el total de palabras en el texto (N).  

Esta probabilidad es la base de los modelos de lenguaje n-gram, porque luego podemos usarla para predecir palabras o analizar patrones de texto.  

En otras palabras: primero entendemos qué tan frecuentes son las palabras por sí solas antes de estudiar cómo se combinan con otras.

# Bigram

Un bigram es una secuencia de dos palabras consecutivas en un texto. En lugar de estudiar palabras de forma individual (como en los unigramas), los bigramas permiten analizar la relación entre pares de palabras, lo que es útil para capturar algo de contexto y patrones en el lenguaje. Por ejemplo, en la frase “el gato duerme”, los bigramas serían: “el gato” y “gato duerme”.  


Un bigram es una secuencia de dos palabras consecutivas en un texto. En lugar de estudiar palabras de forma individual (como en los unigramas), los bigramas permiten analizar la relación entre pares de palabras, lo que es útil para capturar algo de contexto y patrones en el lenguaje. Por ejemplo, en la frase “el gato duerme”, los bigramas serían: “el gato” y “gato duerme”.  

Cómo se calcula:
La probabilidad de que una palabra Wi aparezca dada la palabra anterior Wi-1 se calcula usando la frecuencia de ese par en el texto dividido entre la frecuencia de la primera palabra del par:    

$$P(w_i \mid w_{i-1}) = \frac{\text{count}(w_{i-1}, w_i)}{\text{count}(w_{i-1})}$$

# P(Wi | Wi-1) = count ( Wi-1 n wi ) / count( Wi-1)

Happy book happy chicken

P (book | happy) = count(happy book) / count ( happy)

P (book | happy) = 1 / 2 = 0.5

```python
from nltk.util import ngrams
```

```python
def bigram_probs(initial, words, clean_text, n=2):
    """
    Calcula la probabilidad de que cada palabra de 'words' siga a la palabra 'initial'
    usando bigramas (n=2) en el texto 'clean_text'.
    
    Parámetros:
    initial    : palabra inicial del bigrama
    words      : lista de palabras a evaluar como segunda palabra del bigrama
    clean_text : lista de tokens del texto limpio
    n          : tamaño del n-gram (por defecto 2 para bigramas)
    
    Retorna:
    np.array con probabilidades de cada palabra en 'words' dado 'initial'
    """
    
    # Contar cuántas veces aparece la palabra inicial en el texto
    denominador = len([x for x in clean_text if x == initial])
    
    # Crear la lista de todos los n-gramas del texto
    ngram_list = list(ngrams(clean_text, n))
    
    probs = []  # lista para guardar los conteos/probabilidades
    
    # Para cada palabra candidata en 'words'
    for word in words:
        bigram = (initial, word)  # formar el bigrama
        counter = ngram_list.count(bigram)  # contar cuántas veces aparece
        probs.append(counter)  # guardar el conteo
    
    # Convertir a array y normalizar dividiendo entre la frecuencia de la palabra inicial
    final = np.array(probs) / denominador if denominador > 0 else np.zeros(len(words))
    
    return final

# Ejemplo de uso: calcular probabilidades de bigramas con la palabra "de"
frec2 = bigram_probs("de", words, clean_text, n=2)

# Mostrar DataFrame con palabras y probabilidades
df_bigrams = pd.DataFrame({"palabra": words, "freq": frec2}).sort_values("freq", ascending=False)
print(df_bigrams)

# Verificar que las probabilidades sumen 1 (o cerca)
print("Suma de probabilidades:", df_bigrams["freq"].sum())
```

# Tri Gram

Un trigram es una secuencia de tres palabras consecutivas en un texto. Mientras que los unigramas analizan palabras individuales y los bigramas pares de palabras, los trigramas permiten capturar un contexto más amplio y relaciones entre tres palabras sucesivas, lo que es útil para predecir palabras o analizar patrones de lenguaje. Por ejemplo, en la frase “el gato duerme tranquilo”, los trigramas serían: “el gato duerme” y “gato duerme tranquilo”.  

La probabilidad de que una palabra Wi  aparezca dado su contexto de las dos palabras anteriores (Wi-2,Wi-1)se calcula con la frecuencia del trigram dividido entre la frecuencia del bigrama formado por las dos primeras palabras:  

$$P(w_i \mid w_{i-2}, w_{i-1}) = \frac{\text{count}(w_{i-2}, w_{i-1}, w_i)}{\text{count}(w_{i-2}, w_{i-1})}
$$


# P(Wi | Wi-1 n Wi-2) = count ( Wi-1 n Wi-2 n wi ) / count( Wi-1 n Wi-2)

```python
def trigram_probs(initial_1, initial_2, words, clean_text):
    """
    Calcula la probabilidad condicional de cada palabra dada 
    un contexto de dos palabras anteriores (trigramas).

    Parámetros:
    initial_1 : str - primera palabra del contexto
    initial_2 : str - segunda palabra del contexto
    words     : list - lista de todas las palabras únicas del texto
    clean_text: list - lista de tokens del texto limpio

    Retorna:
    np.array - probabilidades de cada palabra en 'words' dado el contexto
    """

    # Crear lista de todos los trigramas del texto
    trigram_list = list(ngrams(clean_text, 3))

    # Contar cuántas veces aparece el bigrama de contexto (initial_1, initial_2)
    bigram_count = sum(1 for t in trigram_list if t[0] == initial_1 and t[1] == initial_2)

    probs = []  # Lista para guardar las probabilidades de cada palabra

    # Iterar sobre todas las palabras posibles
    for word in words:
        trigram_count = trigram_list.count((initial_1, initial_2, word))  # Contar el trigram específico
        probs.append(trigram_count / bigram_count if bigram_count > 0 else 0)  # Evitar división por cero

    return np.array(probs)


# Ejemplo de uso: calcular probabilidades de trigramas que comienzan con "de mi"
frec3 = trigram_probs("de", "mi", words, clean_text)

# Crear DataFrame para visualizar las palabras y sus probabilidades
df_trigram = pd.DataFrame({"palabra": words, "freq": frec3}).sort_values("freq", ascending=False)
df_trigram
```

## Unimos el proceso

```python
from tqdm import tqdm
```

```python
def sentence(words, clean_text, freq, n=5):
    """
    Genera una oración de longitud 'n' usando probabilidades de unigrama, bigrama y trigramas.

    Parámetros:
    words      : list  - lista de palabras únicas
    clean_text : list  - lista de tokens del texto limpio
    freq       : list  - probabilidades de unigrama de cada palabra
    n          : int   - longitud de la oración a generar (mínimo 2)

    Retorna:
    str - oración generada
    """

    # Paso 1: Elegir la primera palabra usando probabilidad de unigrama
    string1 = np.random.choice(words, p=freq)
    sol = string1  # Inicializamos la oración

    # Paso 2: Elegir la segunda palabra usando probabilidad de bigrama
    bi_prob = bigram_probs(string1, words, clean_text)
    try:
        string2 = np.random.choice(words, p=bi_prob)
    except ValueError:  # si hay problema con probabilidades, elige aleatoria
        string2 = np.random.choice(words)
    sol = sol + " " + string2

    # Paso 3: Elegir las siguientes palabras usando trigramas
    for i in tqdm(range(n-2)):
        tri_prob = trigram_probs(string1, string2, words, clean_text)
        try:
            new_word = np.random.choice(words, p=tri_prob)
        except ValueError:  # si la distribución no es válida, elige aleatoria
            new_word = np.random.choice(words)
        sol = sol + " " + new_word

        # Actualizar contexto para el siguiente trigram
        string1 = string2
        string2 = new_word

    return sol
```

## importancia de texto: TF-IDF

TF-IDF (Term Frequency–Inverse Document Frequency) es una técnica de procesamiento de texto usada en recuperación de información y NLP para cuantificar la importancia de una palabra dentro de un documento en relación con un corpus completo.  

TF (Term Frequency): mide cuántas veces aparece un término en un documento. Cuantas más veces aparece, más relevante puede ser para ese documento.

$$TF(t,d) = \frac{\text{Número de veces que el término } t \text{ aparece en el documento } d}{\text{Número total de términos en el documento } d}
$$

IDF (Inverse Document Frequency): penaliza las palabras que aparecen en muchos documentos, porque son menos informativas (como "el", "de", "y"). Se calcula como:

$$IDF(t, D) = \log \frac{|D|}{1 + |\{d \in D : t \in d\}|}$$

TF-IDF: combina ambos factores:  

$$TF\text{-}IDF(t,d,D) = TF(t,d) \times IDF(t,D)$$

TF-IDF se aplica principalmente sobre colecciones de documentos, donde cada "documento" puede ser:

Una oración

Un párrafo

Un artículo completo

Una página web

Su idea es detectar qué palabras son más representativas de cada unidad de texto en comparación con el resto del corpus. Esto tiene varias aplicaciones:

* Búsqueda y recuperación de información (IR):

        Al indexar un conjunto de documentos, las palabras con TF-IDF alto se consideran más importantes para encontrar documentos relevantes ante una consulta.

        Ejemplo: si buscas “bahía” en tu corpus, los documentos con TF-IDF más alto para “bahía” se muestran primero.

* Extracción de palabras clave:

        Puedes extraer palabras significativas de un texto completo o de un artículo, ignorando palabras comunes como “el”, “de”, “y”.

        TF-IDF permite automáticamente resaltar términos raros pero relevantes.

* Clasificación de documentos:

        Los vectores TF-IDF pueden usarse como features en modelos de machine learning para clasificar textos por tema, sentimiento, o categoría.

* Resúmenes automáticos y análisis de contenido:

        Puedes detectar las palabras más importantes de cada párrafo u oración para generar resúmenes o identificar temas principales.

```python
# Importar la librería
from sklearn.feature_extraction.text import TfidfVectorizer

# Supongamos que tenemos tres documentos de ejemplo
documents = [
    "El gato juega en el jardín",
    "El perro y el gato se llevan bien",
    "El pájaro canta en el jardín"
]

# Crear el vectorizador TF-IDF
vectorizer = TfidfVectorizer()  # Puedes pasar parámetros como stop_words='spanish'

# Ajustar y transformar los documentos a la matriz TF-IDF
tfidf_matrix = vectorizer.fit_transform(documents)

# Obtener los nombres de las palabras (features)
feature_names = vectorizer.get_feature_names_out()

# Convertir la matriz a un DataFrame para visualizarla mejor
import pandas as pd
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print(df_tfidf)
```

Interpretación 

* Valores numéricos: Cada celda en la matriz TF-IDF indica la importancia de una palabra en un documento particular.

        Alto valor → La palabra es muy representativa del documento y no aparece mucho en otros documentos.

        Bajo valor → La palabra es común o no aporta mucha información discriminativa.

* Comparación entre documentos

        Puedes comparar el mismo término en diferentes documentos. Por ejemplo, si la palabra "gato" tiene TF-IDF alto en el primer documento y bajo en los otros, significa que "gato" es distintiva del primer texto.

* Palabras más importantes de un documento

        Para cada fila (documento), puedes ordenar los valores TF-IDF y obtener las palabras más representativas. Estas son útiles para:

                    Resumir contenido.

                    Clasificación de documentos.

                    Extracción de keywords.

* Qué NO indica
        
        TF-IDF no mide el sentimiento ni la relación semántica entre palabras. Solo indica relevancia estadística basada en frecuencia relativa.

```python
# Leer el texto
with open("noche.txt", encoding="utf8") as f:
    text = f.read()

# Dividir en oraciones
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

# Obtener lista de stopwords en español
spanish_stopwords = stopwords.words('spanish')

# Crear vectorizador con stopwords en español
vectorizer = TfidfVectorizer(lowercase=True, stop_words=spanish_stopwords)

# Ajustar y transformar
tfidf_matrix = vectorizer.fit_transform(sentences)

# Convertir a DataFrame para inspeccionar
words = vectorizer.get_feature_names_out()
df = pd.DataFrame(tfidf_matrix.toarray(), columns=words)
print(df)
```

```python

# Leer texto
with open("noche.txt", encoding="utf8") as f:
    text = f.read()

# Dividir en oraciones
sentences = sent_tokenize(text)

# Stopwords en español
spanish_stopwords = stopwords.words('spanish')

# Crear vectorizador TF-IDF
vectorizer = TfidfVectorizer(lowercase=True, stop_words=spanish_stopwords)
tfidf_matrix = vectorizer.fit_transform(sentences)

# Convertir a DataFrame
words = vectorizer.get_feature_names_out()
df = pd.DataFrame(tfidf_matrix.toarray(), columns=words)

# Mostrar las 5 palabras con mayor TF-IDF por oración
for i, sent in enumerate(sentences):
    row = df.iloc[i]
    top_words = row.sort_values(ascending=False).head(5)
    print(f"Oración {i+1}: {sent}")
    print(f"Palabras más representativas: {list(top_words.index)}\n")
```

La importancia de una palabra no solo depende de su frecuencia en la oración, sino también de cuán rara es en todo el corpus. 

* TF (Term Frequency): mide cuántas veces aparece la palabra en la oración.

        Palabras muy repetidas dentro de la oración tienen TF alto.

* IDF (Inverse Document Frequency): mide qué tan rara es la palabra en todo el corpus (todas las oraciones).

        Si una palabra aparece en casi todas las oraciones, su IDF será baja → su TF-IDF baja.

        Si una palabra aparece solo en pocas oraciones, su IDF será alta → su TF-IDF aumenta.

Por lo que, aunque una palabra aparezca una sola vez en la oración, si es rara en el resto del texto, su TF-IDF puede ser alta, lo que indica que es relevante para esa oración.  

Ejemplo simple:  
 
Palabra común: "noche" → aparece en muchas oraciones → TF-IDF baja.

En nuestro ejemplo con “noche.txt”: cada oración es un “documento”. Las palabras con TF-IDF alto en cada oración indican qué términos son clave para esa oración, aunque aparezcan pocas veces en todo el texto.
