# Estadística Inferencial

La estadística inferencial estudia cómo extraer conclusiones generales para toda una población a partir del análisis de una muestra, evaluando el grado de fiabilidad o significación de los resultados obtenidos. Sus principales herramientas son:

- **Muestreo:** Selección de una muestra representativa de la población.
- **Estimación de parámetros:** Determinación de valores aproximados para características poblacionales.
- **Contraste de hipótesis:** Validación o rechazo de afirmaciones sobre la población.

---

## Conceptos Básicos

### Media aritmética
La **media aritmética** es el valor promedio de un conjunto de datos. Se calcula sumando todos los datos y dividiendo entre el número total de elementos. Se representa con la letra griega $\mu$. 

Dada una muestra de $n$ valores $x_i$, la media aritmética se calcula como:

$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

---

### Desviación respecto a la media
La **desviación respecto a la media** es la diferencia en valor absoluto entre cada valor de la variable estadística y la media aritmética:

$$
\text{Desviación}_i = |x_i - \mu|
$$

---

### Varianza
La **varianza** mide la dispersión de los datos y se define como la media aritmética del cuadrado de las desviaciones respecto a la media. Se representa como $\sigma^2$ y se calcula como:

$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
$$

---

### Desviación típica
La **desviación típica** es la raíz cuadrada de la varianza. Es una medida de dispersión más intuitiva, ya que tiene las mismas unidades que los datos originales. Se denota por $\sigma$:

$$
\sigma = \sqrt{\sigma^2}
$$

---

### Moda
La **moda** es el valor que tiene mayor frecuencia absoluta dentro de un conjunto de datos. Se denota como $M_0$.

---

### Mediana
La **mediana** es el valor central de un conjunto de datos ordenados de menor a mayor. Si el número de datos es impar, es el valor único en el centro; si es par, es el promedio de los dos valores centrales. Se representa como:

$$
\tilde{x}
$$

---

### Covarianza
La **covarianza** mide la relación entre dos variables $X$ y $Y$. Se define como la media de los productos de las desviaciones de cada variable respecto a sus medias. Se representa como $\sigma_{xy}$ y se calcula como:

$$
\sigma_{xy} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu_x)(y_i - \mu_y)
$$

- Si $\sigma_{xy} > 0$, la correlación es directa.
- Si $\sigma_{xy} < 0$, la correlación es inversa.

---

### Correlación
La **correlación** cuantifica la relación o dependencia entre dos variables. Se utiliza el coeficiente de correlación de Pearson ($r$) para medir esta relación. Su fórmula es:

$$
r = \frac{\sigma_{xy}}{\sigma_x \sigma_y}
$$

- **$r \in [-1,1] $:**

---

### Valores atípicos
Un **valor atípico** es una observación que se aleja significativamente del resto de los datos. Puede ser causado por errores en la recolección de datos o por fenómenos inusuales. Por lo general, se recomienda eliminarlos del análisis si afectan los resultados.

---

**Nota:** Estos conceptos básicos son esenciales para comprender los métodos de la estadística inferencial y aplicarlos adecuadamente en análisis de datos.

# Boxplots: Explicación

Un **boxplot** (o diagrama de cajas y bigotes) es una herramienta gráfica utilizada para representar la distribución de un conjunto de datos. Es particularmente útil para detectar valores atípicos y comprender la dispersión y simetría de los datos.

## Elementos de un Boxplot

1. **Caja (Box)**:
   - Representa el rango intercuartil (IQR, por sus siglas en inglés), que abarca del primer cuartil ($Q_1$) al tercer cuartil ($Q_3$).
   - El IQR contiene el 50% de los datos centrales.

2. **Línea dentro de la caja**:
   - Indica la **mediana** ($Q_2$) del conjunto de datos. Divide los datos en dos mitades iguales.

3. **Bigotes (Whiskers)**:
   - Se extienden hasta los valores más lejanos dentro de 1.5 veces el rango intercuartil desde los cuartiles ($Q_1 - 1.5 \cdot IQR$ y $Q_3 + 1.5 \cdot IQR$).
   - Representan los límites para los valores considerados no atípicos.

4. **Puntos fuera de los bigotes**:
   - Representan **valores atípicos** (outliers), es decir, aquellos datos que caen fuera del rango establecido por los bigotes.

## Interpretación
El boxplot facilita la comprensión de:
- **Mediana:** Valor central de los datos.
- **Rango intercuartil:** Medida de dispersión que indica la concentración de datos.
- **Simetría:** Si la caja está centrada respecto a los bigotes, los datos son aproximadamente simétricos.
- **Valores atípicos:** Identifica observaciones que se alejan significativamente del resto.

---

# Curtosis y Exceso de Curtosis

La **curtosis** mide el grado de concentración de los datos en las colas de una distribución. Ayuda a evaluar si los datos tienen colas más gruesas o delgadas en comparación con una distribución normal.

## Tipos de Distribuciones Basados en la Curtosis

1. **Platicúrtica**:
   - Si $\text{Curtosis} < 3$.
   - Las colas son más delgadas que las de una distribución normal.
   - Ejemplo: Distribuciones más planas.

2. **Mesocúrtica**:
   - Si $\text{Curtosis} = 3$.
   - Las colas tienen un grosor similar al de una distribución normal.
   - Ejemplo: La distribución normal es mesocúrtica.

3. **Leptocúrtica**:
   - Si $\text{Curtosis} > 3$.
   - Las colas son más gruesas que las de una distribución normal.
   - Ejemplo: Distribuciones con más concentración de datos en las colas.

---

## Exceso de Curtosis

El **exceso de curtosis** se calcula como:

$$
\text{Exceso de Curtosis} = \text{Curtosis} - 3
$$

Esto ajusta la medida para centrarse en la comparación con la curtosis de una distribución normal.

### Interpretación del Exceso de Curtosis

1. **Platicúrtica**:
   - Si $\text{Exceso de Curtosis} < 0$.
   - Colas más delgadas que una distribución normal.

2. **Mesocúrtica**:
   - Si $\text{Exceso de Curtosis} = 0$.
   - Colas similares a una distribución normal.

3. **Leptocúrtica**:
   - Si $\text{Exceso de Curtosis} > 0$.
   - Colas más gruesas que una distribución normal.

---

## Fórmula General de la Curtosis

La curtosis se calcula como:

$$
\text{Curtosis} = \frac{n \sum (x_i - \bar{x})^4}{\left[ \sum (x_i - \bar{x})^2 \right]^2} - 3
$$

Donde:
- $n$ es el número de datos.
- $x_i$ son los valores de la variable.
- $\bar{x}$ es la media de los datos.

El ajuste de `-3` en esta fórmula elimina el valor esperado de la curtosis para una distribución normal, de manera que la curtosis sea cero en el caso mesocúrtico.

---

## Nota Importante

Ambas definiciones (curtosis estándar y exceso de curtosis) son equivalentes para determinar si una distribución es platicúrtica, mesocúrtica o leptocúrtica, pero se deben aclarar en contexto según el análisis realizado.

# KDE: Kernel Density Estimation (Estimación de Densidad por Núcleo)

El **KDE** es una técnica estadística utilizada para estimar la **función de densidad de probabilidad** (PDF, por sus siglas en inglés) de una variable continua. En otras palabras, KDE suaviza los datos y muestra una curva que representa la distribución subyacente de los mismos.

---

## ¿Cómo funciona el KDE?

1. **Idea básica**:
   - En lugar de representar los datos mediante un histograma, que agrupa los valores en intervalos (bins), el KDE coloca un "núcleo" (kernel) en cada dato individual.
   - Estos núcleos se combinan para crear una curva continua que estima la densidad de probabilidad.

2. **Núcleo (Kernel)**:
   - Es una función matemática (como una gaussiana) que se centra en cada dato.
   - Tiene un parámetro llamado **ancho de banda (bandwidth)**, que determina qué tan ancha o estrecha será la curva en cada punto.

3. **Suma de núcleos**:
   - La curva final es la suma de todos los núcleos aplicados a los datos, lo que produce una estimación suave de la distribución.

---

## Parámetros Clave

1. **Ancho de Banda (Bandwidth)**:
   - Determina qué tan suave o detallada será la curva.
   - Un ancho de banda más pequeño crea curvas más ajustadas a los datos (puede sobreajustarse).
   - Un ancho de banda más grande genera curvas más suaves, pero puede perder detalles.

2. **Función Kernel**:
   - Define la forma del núcleo aplicado a cada dato.
   - Ejemplos comunes de funciones kernel:
     - Gaussiana.
     - Uniforme.
     - Epanechnikov.

---

## Comparación: Histograma vs KDE

| **Aspecto**          | **Histograma**                            | **KDE**                                     |
|-----------------------|-------------------------------------------|---------------------------------------------|
| **Representación**    | Agrupa datos en intervalos (bins).        | Suaviza los datos para una curva continua.  |
| **Parámetros**        | Número de bins.                          | Ancho de banda y tipo de kernel.            |
| **Resultados**        | Gráfico discreto (barras).               | Gráfico continuo (curva).                   |
| **Flexibilidad**      | Depende del número de bins.              | Más flexible gracias al ancho de banda.     |

---

## Ejemplo Práctico
Si tenemos los datos de edades de los pasajeros del Titanic, el KDE nos permite visualizar una curva continua que representa cómo están distribuidas las edades en lugar de usar un histograma de barras.

---

## Ventajas del KDE
- Proporciona una representación más suave y continua de la distribución.
- No depende de la elección de bins como en un histograma.
- Ideal para datos con distribuciones complejas.

## Limitaciones
- Puede ser sensible al ancho de banda: un valor incorrecto puede producir sobreajuste o subajuste.
- Es computacionalmente más intensivo que un histograma.

En resumen, el KDE es una poderosa herramienta para explorar y visualizar la distribución de datos continuos.


```python
# Histograma acumulativo con curva de densidad
plt.figure(figsize=(10, 6))
sns.histplot(poblacion["Age"].dropna(), bins=30, cumulative=True, kde=True, color="blue", alpha=0.6)
plt.title("Histograma Acumulativo de Edades con KDE", fontsize=14)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Frecuencia Acumulada", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Histograma clásico con curva de densidad
plt.figure(figsize=(10, 6))
sns.histplot(poblacion["Age"].dropna(), bins=30, cumulative=False, kde=True, color="green", alpha=0.6)
plt.title("Histograma de Edades con KDE", fontsize=14)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Frecuencia Relativa", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Graficar KDE con diferentes valores de ancho de banda
plt.figure(figsize=(12, 6))

# KDE con ancho de banda estándar
sns.kdeplot(poblacion["Age"].dropna(), bw_adjust=1, label="Ancho estándar (bw_adjust=1)", color="blue",kernel = "epa")

# KDE con ancho de banda más pequeño (curva más ajustada)
sns.kdeplot(poblacion["Age"].dropna(), bw_adjust=0.5, label="Ancho reducido (bw_adjust=0.5)", color="red")

# KDE con ancho de banda más grande (curva más suave)
sns.kdeplot(poblacion["Age"].dropna(), bw_adjust=2, label="Ancho aumentado (bw_adjust=2)", color="green")

# Configuración del gráfico
plt.title("Efecto del Ancho del Kernel en el KDE", fontsize=14)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# Tipos de Muestreo

El muestreo es el proceso de seleccionar una parte representativa de una población para realizar análisis estadísticos. Dentro de los tipos de muestreo, el **muestreo aleatorio simple** es uno de los métodos más utilizados.

---

## Muestreo Aleatorio Simple

El **muestreo aleatorio simple** consiste en seleccionar una muestra de una población de forma que **todas las muestras del mismo tamaño tengan la misma probabilidad de ser elegidas**. Es decir, cada individuo de la población tiene una probabilidad uniforme de ser seleccionado.

### Características Principales
- Cada elemento de la población tiene la misma probabilidad de ser seleccionado.
- Es un método sencillo y no sesgado si se implementa correctamente.
- Requiere que todos los individuos de la población estén identificados (por ejemplo, con un listado).

### Escenarios Aplicables
El muestreo aleatorio simple es útil cuando:
- La población es homogénea.
- Es posible acceder a todos los elementos de la población.
- Se requiere un análisis no sesgado de la población.

---

### Tipos de Muestreo Aleatorio

#### 1. Muestreo con Reemplazo
- Cada elemento seleccionado **puede volver a ser seleccionado** en iteraciones futuras.
- Después de seleccionar un elemento, este se **devuelve a la población**.
- **Ventaja**: Se puede generar un número infinito de muestras posibles.
- **Escenario**: Útil cuando la población es pequeña pero se necesita una gran cantidad de muestras.

#### 2. Muestreo Sin Reemplazo
- Cada elemento seleccionado **no puede volver a ser seleccionado**.
- Después de seleccionar un elemento, este se **elimina de la población**.
- **Ventaja**: No hay repeticiones, por lo que cada muestra es única.
- **Escenario**: Comúnmente utilizado cuando el objetivo es estudiar una proporción específica o asegurar que cada individuo sea evaluado una sola vez.

---

### Ejemplo Práctico
Supongamos que queremos seleccionar estudiantes de una clase para un experimento:

1. **Con Reemplazo**:
   - Elegimos un estudiante al azar, pero ese estudiante **puede ser seleccionado de nuevo** en una iteración posterior.
   - Ejemplo: Elegimos tres estudiantes, y el resultado podría ser: [A, B, A].

2. **Sin Reemplazo**:
   - Elegimos un estudiante al azar, y una vez seleccionado, **no puede ser elegido nuevamente**.
   - Ejemplo: Elegimos tres estudiantes, y el resultado podría ser: [A, B, C].

El tipo de muestreo elegido dependerá de las características del experimento y de la población a estudiar.

# Muestreo Sistemático (Isométrico)

El **muestreo sistemático** (también llamado muestreo isométrico) es un método en el que los elementos se seleccionan a intervalos regulares de una población ordenada. Este método requiere numerar a los sujetos de la población en orden, calcular un **intervalo de muestreo** fijo, y extraer muestras según este intervalo.

---

## ¿Cómo funciona?

1. **Numerar los sujetos**:
   - Asignar un número único a cada sujeto de la población.

2. **Calcular el intervalo de muestreo**:
   - Si el tamaño de la población es $N$ y el tamaño deseado de la muestra es $n$, el intervalo de muestreo ($k$) se calcula como:
     $$
     k = \frac{N}{n}
     $$

3. **Elegir un punto inicial**:
   - Seleccionar aleatoriamente un sujeto como punto de partida dentro del primer intervalo.

4. **Seleccionar sujetos**:
   - Elegir los sujetos cada $k$ posiciones a partir del punto inicial.

---

## Ejemplo Práctico

Supongamos:
- Una población de $N = 100$ elementos.
- Se desea una muestra de $n = 10$ elementos.

### Pasos:
1. **Calcular el intervalo**:
   $$
   k = \frac{100}{10} = 10
   $$

2. **Elegir un punto inicial**:
   - Seleccionamos aleatoriamente un número del 1 al 10, por ejemplo, el 3.

3. **Seleccionar los sujetos**:
   - Los sujetos seleccionados serán: 3, 13, 23, 33, 43, 53, 63, 73, 83, 93.

---

## Defectos

El muestreo sistemático tiene limitaciones cuando los datos presentan:
1. **Tendencias crecientes o decrecientes**:
   - Si la población sigue una tendencia, los intervalos pueden sesgar la muestra.
2. **Leyes periódicas**:
   - Si la población tiene una estructura periódica, el muestreo sistemático puede seleccionar muestras que no representen adecuadamente la población.

---

## Escenarios Aplicables

El muestreo sistemático es ideal cuando:
1. Los individuos están distribuidos **uniformemente**.
2. No hay tendencias evidentes (crecientes o decrecientes) en la población.

# Muestreo Estratificado

El **muestreo estratificado** es un método que divide la población en **estratos** o subgrupos basados en características específicas (como atributos, categorías o etiquetas). Luego, se seleccionan muestras de cada estrato mediante métodos de muestreo, como aleatorio simple o sistemático, para garantizar que cada subgrupo esté representado en la muestra final.

---

## ¿Cómo Funciona?

1. **Dividir en Estratos**:
   - Identificar características clave que diferencian a los individuos de la población.
   - Dividir la población en subgrupos homogéneos (estratos) según esas características.

2. **Muestrear Dentro de Cada Estrato**:
   - Aplicar un método de muestreo (como aleatorio simple o sistemático) para seleccionar elementos de cada estrato.
   - El tamaño de la muestra puede ser proporcional al tamaño del estrato o igual para todos los estratos, dependiendo del objetivo del estudio.

3. **Combinar las Muestras**:
   - Unir las muestras seleccionadas de cada estrato para formar la muestra final.

---

## Ventajas del Muestreo Estratificado

1. **Reducción del Error de Muestreo**:
   - Al dividir la población en estratos homogéneos, se reduce la variabilidad dentro de cada grupo, mejorando la precisión del muestreo.

2. **Representación de Subgrupos**:
   - Garantiza que todos los estratos estén representados en la muestra, incluso aquellos que son pequeños o menos comunes.

3. **Análisis Específico**:
   - Permite analizar resultados por separado para cada estrato, lo que es útil en estudios donde las características de los subgrupos son importantes.

---

## Escenarios Aplicables

El muestreo estratificado es ideal para datos que:
- Contienen atributos o etiquetas que permiten clasificarlos en subgrupos homogéneos.
- Requieren representación equitativa o proporcional de diferentes categorías.
- Incluyen subgrupos que podrían ser ignorados en un muestreo aleatorio simple.

Ejemplos:
- Encuestas de opinión donde se quiere asegurar representación por género, edad o región.
- Estudios académicos donde los estudiantes se agrupan por nivel educativo.
- Análisis de mercado donde los clientes se estratifican por nivel de ingresos.

---

# Muestreo por Conglomerados

El **muestreo por conglomerados** es un método de muestreo probabilístico en el que se seleccionan subconjuntos (o conglomerados) de la población en lugar de elementos individuales. Cada conglomerado se trata como una unidad de muestreo, y todos los elementos dentro de los conglomerados seleccionados forman parte de la muestra final.

---

## ¿Cómo Funciona?

1. **Dividir la Población en Conglomerados**:
   - La población completa se divide en subconjuntos o conglomerados, idealmente homogéneos internamente y representativos de la población total.

2. **Seleccionar Conglomerados Aleatoriamente**:
   - Se selecciona una cantidad específica de conglomerados mediante un método probabilístico, como el muestreo aleatorio simple.

3. **Incluir Todos los Elementos del Conglomerado**:
   - Todos los individuos dentro de los conglomerados seleccionados forman parte de la muestra.

---

## Ventajas del Muestreo por Conglomerados

1. **Eficiencia Logística**:
   - Reduce costos y tiempo, ya que la recolección de datos se limita a los conglomerados seleccionados.
   
2. **Practicidad**:
   - Es útil cuando la población está geográficamente dispersa, lo que hace difícil implementar otros tipos de muestreo.

3. **Fácil Implementación**:
   - Solo es necesario identificar y muestrear conglomerados en lugar de individuos.

---

## Desventajas del Muestreo por Conglomerados

1. **Errores de Muestreo**:
   - Si los conglomerados no son representativos de la población total, se pueden introducir errores significativos.
   - Los conglomerados más pequeños tienden a generar mayor error.

2. **Dependencia de la Calidad de la División**:
   - Los conglomerados deben ser homogéneos internamente y similares entre sí. Si los conglomerados están mal definidos, el muestreo será sesgado.

---

## Escenarios Aplicables

El muestreo por conglomerados es ideal cuando:
- La población está organizada naturalmente en grupos o regiones (por ejemplo, escuelas, ciudades o distritos).
- Es difícil o costoso acceder a los individuos directamente.
- Los conglomerados tienen características similares y son representativos de la población total.

### Condición Ideal:
Los conglomerados deben ser:
- **Homogéneos internamente**: Elementos dentro del mismo conglomerado deben ser similares.
- **Representativos entre sí**: Cada conglomerado debe ser una minirrepresentación de toda la población.

---

## Ejemplo Práctico

Supongamos que queremos estudiar el desempeño académico de los estudiantes en un país con 1,000 escuelas. En lugar de seleccionar estudiantes individuales, aplicamos el muestreo por conglomerados:

1. **Definir los Conglomerados**:
   - Cada escuela es un conglomerado.

2. **Seleccionar los Conglomerados**:
   - Seleccionamos aleatoriamente 50 escuelas.

3. **Recolectar Datos**:
   - Incluimos a todos los estudiantes de las 50 escuelas seleccionadas en la muestra final.

---

## Comparación: Muestreo por Conglomerados vs. Muestreo Estratificado

| **Aspecto**                 | **Muestreo Estratificado**                             | **Muestreo por Conglomerados**                 |
|-----------------------------|-------------------------------------------------------|-----------------------------------------------|
| **Definición**              | Divide en estratos homogéneos, luego selecciona muestras dentro de cada estrato. | Divide en conglomerados, luego selecciona conglomerados completos. |
| **Homogeneidad**            | Estratos homogéneos internamente, diferentes entre sí. | Conglomerados homogéneos entre sí, heterogéneos internamente. |
| **Representación**          | Asegura la representación proporcional de cada subgrupo. | Se seleccionan conglomerados enteros.         |

El **muestreo por conglomerados** es útil para poblaciones grandes y dispersas, siempre que los conglomerados estén bien definidos y sean representativos.


# Diferencia entre Estrato y Conglomerado

La diferencia principal entre un **estrato** y un **conglomerado** radica en cómo se agrupan los elementos de la población y cómo se seleccionan para formar una muestra.

---

## 1. Estrato

Un **estrato** es un subgrupo homogéneo dentro de una población, formado según una característica específica.

### Características:
- **Homogeneidad interna**: Los elementos dentro de un estrato son similares en cuanto a la característica que define al estrato.
  - Ejemplo: Género, nivel educativo, región, etc.
- **Diferencias entre estratos**: Los estratos son distintos entre sí.
  - Ejemplo: En una población dividida por género, los estratos serían "masculino" y "femenino".
- Se seleccionan muestras de **todos los estratos**, garantizando que cada uno esté representado en la muestra final.

### Ejemplo:
En una encuesta nacional, los **estratos** podrían ser:
- Regiones geográficas (Norte, Centro, Sur).
- Grupos de edad (jóvenes, adultos, mayores).
En este caso, se seleccionan individuos de cada estrato para formar una muestra representativa.

---

## 2. Conglomerado

Un **conglomerado** es un subconjunto de la población que funciona como una mini-representación de la población completa.

### Características:
- **Heterogeneidad interna**: Los elementos dentro de un conglomerado son diversos y representan características de toda la población.
  - Ejemplo: Una escuela puede representar una mezcla de estudiantes con diferentes géneros, edades y niveles educativos.
- **Similitud entre conglomerados**: Los conglomerados deben ser lo más similares posible entre sí para evitar sesgo.
  - Ejemplo: Todas las escuelas deberían ser comparables en estructura y composición.
- Se seleccionan **algunos conglomerados completos**, y no necesariamente todos.

### Ejemplo:
En una encuesta educativa, los **conglomerados** podrían ser:
- Escuelas.
- Ciudades.
En este caso, se seleccionan algunas escuelas o ciudades completas para recolectar datos de todos sus integrantes.

---

## Diferencias Clave

| **Aspecto**            | **Estrato**                                     | **Conglomerado**                                 |
|-------------------------|------------------------------------------------|-------------------------------------------------|
| **Definición**          | Subgrupo homogéneo dentro de la población.      | Subconjunto heterogéneo que representa a la población. |
| **Propósito**           | Garantizar representación de cada subgrupo.    | Facilitar el muestreo mediante selección de grupos completos. |
| **Homogeneidad**        | Homogéneos dentro del estrato, diferentes entre sí. | Heterogéneos dentro del conglomerado, similares entre sí. |
| **Selección**           | Se seleccionan muestras de cada estrato.        | Se seleccionan algunos conglomerados completos. |
| **Escenarios Aplicables** | Cuando se conocen características clave de los individuos. | Cuando la población está organizada en grupos naturales. |
