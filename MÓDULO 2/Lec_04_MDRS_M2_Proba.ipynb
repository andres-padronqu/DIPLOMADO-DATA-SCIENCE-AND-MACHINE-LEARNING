$ P(A) = \frac{\#\;resultados\;favorables\;A}{\#\;resultados\;totales\;\Omega} $

$ P(E) = lim_{n->\infty} \frac{n(E)}{n} $

```python
n = 5_000 #numero de puntos aleatorios
a, b = 0,1
```

```python
# Generamos numeros aleatorios

x_rand = np.random.uniform(a,b,n)
y_rand = np.random.uniform(0,1,n)
```

```python
f = lambda x: x**2

# Puntos po debajo y por encima de la curva

under_curve = y_rand <= f(x_rand)
over_curve = ~under_curve
```

```python
# estimar montecarlo

area_rect = (b-a)*1

proportion = np.sum(under_curve)/n

integral_estimate = area_rect *proportion
```

```python
# grafico

plt.figure(figsize=(10,10))
plt.scatter(x_rand[under_curve], y_rand[under_curve], color="green", s=5, label="Debajo de la curva")
plt.scatter(x_rand[over_curve], y_rand[over_curve], color="red", s=5, label="Sobre de la curva")
x_vals = np.linspace(a, b, 500)
plt.plot(x_vals, f(x_vals), color= "blue", linewidth=2)
plt.title(f"INTEGRACION POR MC (Aprox: {integral_estimate:.4f}, Exacta: {1/3}:4.f)")

```

### Probabilidad condicional ###
La probabilidad condicional generalmente se utiliza para examinar eventos en subconjuntos del espacio muestral. Específicamente, estamos interesados en la probabilidad de que un evento ocurra solo si otro evento ha ocurrido primero. Por ejemplo: la probabilidad de que una institución financiera quiebre dado que otra institución financiera ha quebrado.  
$ P(A|B)=\frac{P(A\cap B)}{P(B)} $  



La probabilidad condicional nos permite medir la probabilidad de que ocurra un evento considerando que ha ocurrido otro anteriormente. 

Ejemplo: Se considera el experimento de lanzar un dado equilibrado.  
A = {2} = Se obtiene el número 2  
B = {2,4,6} = se obtiene un par  

Es claro que P(A) = 1/6, sin embargo sabiendo que B ha ocurrido, la Proba de A ahora es  
$P(A|B)=\frac{P(A\cap B)}{P(B)}  = \frac{P(2)}{P(2,4,6)} = \frac{1/6}{1/2} = 1/3$

### Probabilidad Total

Sea $\Omega$ un espacio muestral de un experimento aleatorio, decimos que la coleccion de eventos {B1,B2,...,Bn} es una particion finita de $\Omega$ si  
1.- Bi es diferente del vacio  
2.- La intersección de cualquier par de Bi´s es igual al vacio  
3.- La unión de todos los Bi´s es $\Omega$  

Teorema  
Sea B1,...Bn una partición de $\Omega$, tal que la probabilidad de cada Bi es diferente de 0 para todo i, entonces para cualquier evento A se tiene:  
$P(A) =\sum_{i=1}^{n}P(A|B_{i})P(B_{i})$

Ejemplo: Supon que en una población de igual número de hombres y mujeres, el 4% de hombres tienen covid y el 1 % de mujeres tienen covid.  Si una persona es seleccionada al azar, cual es la probabilidad de que tenga covid?  

M: Es mujer  
H: Es hombre  
C = Tiene covid  

$P(C) = P(C|M)P(M)+ P(C|H)P(H)$
