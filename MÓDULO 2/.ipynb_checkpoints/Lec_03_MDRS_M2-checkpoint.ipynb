### Eigendescomposición ###
Definida solo para matrices cuadradas. El objetivo es extraer dos sets de features de una matriz: sus eigenvalores y eigenvectores

**Ecuación fundamental de eigenvalores:**  
**$Av=\lambda v$**  

Para obtener los eigenvalores y eigenvectores se puede utilizar la ecuación característica:  
$ (A-\lambda I)v=0 $  
$ det(A-\lambda I)=0 $

```python
# Definimos la matriz A
A = np.array([[4, 1], [2, 3]])

# Calculamos los eigenvalores y eigenvectores
eigenvalores, eigenvectores = np.linalg.eig(A)

# Mostramos los resultados
print("Eigenvalores:")
print(eigenvalores)

print("Eigenvectores:") #no son unicos y python resuelve sistemas ecs de distinta manera
print(eigenvectores)

# Gráfica de los eigenvectores
origin = [0, 0]  # El origen de los vectores

plt.figure(figsize=(6, 6))
plt.quiver(*origin, eigenvectores[0, 0], eigenvectores[1, 0], angles='xy', scale_units='xy', scale=1, color='r', label='Eigenvector 1')
plt.quiver(*origin, eigenvectores[0, 1], eigenvectores[1, 1], angles='xy', scale_units='xy', scale=1, color='b', label='Eigenvector 2')

# Configuración del gráfico
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title('Eigenvectores de la matriz A')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar el gráfico
plt.show()
```

# PCA

**Principal Components Analysis (PCA)**  
El objetivo del PCA es identificar una base tal que cada vector que conforma la base maximiza la covarianza manteniendo la ortogonalidad a todos los demás vectores que conforman la base.  

Es un método que sirve para reducir la dimensionalidad de los datos, podemos pensarlo como si proyectaramos una matriz con $m$ columnas en un subespacio con $m$ o menos columnas manteniendo la esencia de los datos originales.  

1. Calculamos el promedio de cada columna.
2. Centramos los valores de cada columna.
3. Calculamos la matriz de covarianza de los datos centrados.
4. Calculamos la eigendescomposición de la matriz de covarianza.

```python
# Creamos un conjunto de datos de ejemplo (10 muestras, 2 variables)
X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])

# Estandarizamos los datos (media 0, desviación estándar 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicamos PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Mostramos los componentes principales
print("Componentes principales:")
print(pca.components_)

print("\nVarianza explicada por cada componente:")
print(pca.explained_variance_ratio_)

# Visualizamos los datos originales y los componentes principales
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], label="Datos Originales", color='blue')

# Dibujamos los vectores de los componentes principales
for length, vector in zip(pca.explained_variance_ratio_, pca.components_):
    v = vector * 3 * np.sqrt(length)  # Escalar los componentes principales
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red')

plt.title("PCA: Componentes principales")
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.grid(True)
plt.legend()
plt.show()
```

# Reducción de Dimensionalidad con PCA

```python
# Datos de ejemplo (10 muestras, 2 variables)
X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])

# Estandarización de los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicamos PCA para reducir a 1 componente principal
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

# Reconstrucción de los datos en el espacio original (1D a 2D)
X_reconstructed = pca.inverse_transform(X_pca)

# Mostramos los componentes principales y la varianza explicada
print("Componentes principales:")
print(pca.components_)

print("\nVarianza explicada por cada componente:")
print(pca.explained_variance_ratio_)

# Visualización
plt.figure(figsize=(10, 7))

# Datos originales
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='blue', label='Datos Originales')

# Componentes principales
for length, vector in zip(pca.explained_variance_ratio_, pca.components_):
    v = vector * 3 * np.sqrt(length)  # Escalar los componentes principales
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Componente Principal 1')

# Datos proyectados
plt.scatter(X_reconstructed[:, 0], X_reconstructed[:, 1], color='green', label='Datos Reconstruidos (1D)')

# Líneas de proyección
for i in range(len(X_scaled)):
    plt.plot([X_scaled[i, 0], X_reconstructed[i, 0]], [X_scaled[i, 1], X_reconstructed[i, 1]], 'k--', linewidth=0.5)

plt.title("PCA: Reducción de Dimensionalidad de 2D a 1D")
plt.xlabel("Variable 1 (Estandarizada)")
plt.ylabel("Variable 2 (Estandarizada)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def estandariza_datos(datos):
    '''Función para estandarizar los datos de un numpy array
    ENTRADA: 
    datos: numpy array (matriz)
    
    SALIDA: 
    array con datos estandarizados (media cero y desviación estándar 1)
    
    
    '''
    
    #Se obtiene la media de cada columna
    medias=datos.mean(axis=0)
    
    #desviación estándar
    desv=datos.std(axis=0)
    
    #Se normalizan los datos 
    datos_norm=(datos-medias)/desv
    
    return datos_norm
```

```python
def pca(datos_norm):
    '''Función realizar el PCA
    
    ENTRADA: 
    datos_norm: Numpy array (matriz) de datos normalizados
    
    SALIDA: 
    mat_comp_prin: numpy array que representa la matriz de componentes
    principales
    
    eig_vect: numpy array que represetna la amtriz cuya i-ésima columna es
    el vector característico correspondiente al i-ésimo valor característico
    
    eig_vals: numpy array que contiene los valores característicos 
    ordendos de myor a menor

    
    '''
        #Calculamos la matrzi de correlaciones 
    mat_cor=np.corrcoef(datos_norm,rowvar=False)
    
    #Calculamos los valores y vectores característicos de la matriz
    #de correlaciones (np.linalg.eig)
    eig_vals, eig_vect=np.linalg.eig(mat_cor)
    
    #Ordena los valores característicos de mayor a menor
    indices_orden=eig_vals.argsort()[::-1]
    eig_vals=eig_vals[indices_orden]
    
    #ordenamos las columnas
    eig_vect=eig_vect[:,indices_orden]
    
    #calculamos la matriz de componentes principales. 
    mat_compo_prin=np.matmul(datos_norm,eig_vect)
    
    return mat_compo_prin, eig_vect, eig_vals
```

```python
#Estandarizamos los datos
datos_norm=estandariza_datos(datos)
mat_comp_prin,eig_vector,eig_vals=pca(datos_norm)
```

```python
#Vamos a graficrlos. 
def base_100(serie):
    '''Función para obtener un índice base 100 dada una serie
    ENTRADA: 
    serie: numpy array de dimensión 1
    
    SALIDA: 
    numpy array de dimensión 1 cuya primera entrada es igual a 100'''
    
    x0=100
    p0=serie[0]
    
    return (serie / p0) * x0
```

```python
# Función base 100 (normalización)
def base_100(serie):
    x0 = 100
    p0 = serie[0]
    return (serie / p0) * x0

# Cantidad de índices
n_indices = datos.shape[1]

# Nombres de los índices
nombres_indices = ["DIA (Dow Jones)", "EWW (México)", "QQQ (Nasdaq)", "IWM (Russell 2000)", "SPY (S&P 500)"]

#Paleta de colores
colores = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]

# Tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficamos cada índice
for i in range(n_indices):
    indice = base_100(datos_norm[:, i])
    plt.plot(indice, color=colores[i], label=nombres_indices[i])

# Graficamos el primer componente principal
indice_pca = base_100(mat_comp_prin[:, 0])
plt.plot(indice_pca, color="blue", marker='o', linestyle='dashed',
         linewidth=1, markersize=4, label="Primer Componente Principal (PCA)")

# Añadimos título y etiquetas
plt.title("Comparación de Índices y Primer Componente Principal", fontsize=14)
plt.xlabel("Observaciones", fontsize=12)
plt.ylabel("Índice (Base 100)", fontsize=12)

# Añadimos una leyenda
plt.legend(loc="best")

# Añadimos una cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

# Mostramos el gráfico
plt.tight_layout()
plt.show()
```
