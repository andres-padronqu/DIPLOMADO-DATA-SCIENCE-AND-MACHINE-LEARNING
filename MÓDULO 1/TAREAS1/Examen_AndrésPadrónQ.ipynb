### Problema 1

Creamos una función que recibe un texto y devuelve las 10 palabras que más se repiten.

**Pasos que seguimos:**

1. Convertimos todo a minúsculas.
2. Quitamos signos de puntuación comunes manualmente con `.replace()`.
3. Partimos el texto con `.split()` para obtener las palabras.
4. Contamos cuántas veces aparece cada palabra usando un diccionario.
5. Ordenamos las palabras por su frecuencia y devolvemos las 10 primeras.



````python
# Creamos la función que contabilice las palabras más frecuentes
def palabras_mas_frecuentes(texto):
    
    # Convertimos todo a minúsculas
    texto = texto.lower()
    
    # Quitamos signos de puntuación
    for simbolo in [".", ",", "¿", "?", "¡", "!", "(", ")", ":", ";", '"']:
        texto = texto.replace(simbolo, "")
    
    # Separamos las palabras del texto
    palabras = texto.split()
    
    # Contamos ocurrencias usando un diccionario
    conteo = {}
    for palabra in palabras:
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1
    
    # Ordenamos por frecuencia (de mayor a menor)
    lista_ordenada = sorted(conteo.items(), key=lambda x: x[1], reverse=True)
    
    # Tomamos las 10 palabras más repetidas
    top_10 = lista_ordenada[:10]
    
    # Convertimos de nuevo a diccionario
    resultado = {}
    for palabra, cantidad in top_10:
        resultado[palabra] = cantidad
    
    return resultado
````

## Problema 3

**Descripción:**

Creamos una función que recibe un punto $A = (x, y)$ y devuelve sus coordenadas polares $(r, \theta)$.

Usamos `math.sqrt()` para calcular el radio y `math.atan2()` para obtener el ángulo considerando el cuadrante.

Si el ángulo es negativo, le sumamos $2\pi$ para ajustarlo al intervalo $[0,\ 2\pi)$.


````python
# Importamos la librería math para usar funciones trigonométricas y raíz cuadrada
import math

# Definimos una función que recibe un punto A en coordenadas rectangulares (x, y)

def coord_rect_a_polares(A):
    x = A[0]  # Extraemos la coordenada x del punto
    y = A[1]  # Extraemos la coordenada y del punto

    # Calculamos el radio con la fórmula: r = sqrt(x^2 + y^2)
    r = math.sqrt(x**2 + y**2)

    # Calculamos el ángulo theta usando atan2 para obtener el ángulo correcto según el cuadrante
    theta = math.atan2(y, x)  # atan2 ya devuelve el ángulo en radianes entre [-π, π]

    # Si el ángulo es negativo, lo convertimos al rango [0, 2π] sumando 2π
    if theta < 0:
        theta += 2 * math.pi

    # Devolvemos el resultado como una tupla: (r, theta)
    return (r, theta)
````

````python
A = (2, 2)  # Punto en coordenadas rectangulares
print(coord_rect_a_polares(A))
````

## Problema 4

**Descripción:**  
Simulamos `n` puntos uniformes en el cuadrado $[0,1] \times [0,1]$, y contamos cuántos caen dentro del cuarto de círculo (cuando $x^2 + y^2 \leq 1$).

Luego usamos la fórmula de Monte Carlo:

$$
\pi \approx 4 \cdot \frac{\text{# puntos dentro del círculo}}{\text{# puntos totales}}
$$

Mientras más puntos simules, más precisa será la aproximación de $\pi$.


````python
# Importamos Numpy
import numpy as np

# Definimos una función que aproxime pi con n números
def approximate_pi(n):
    
    # Generamos n valores uniformes en [0,1] para x
    x = np.random.uniform(0, 1, n)
    
    # Generamos n valores uniformes en [0,1] para y
    y = np.random.uniform(0, 1, n)
    
    # Calculamos x^2 + y^2 para cada par (x, y)
    distancia_cuadrada = x**2 + y**2
    
    # Contamos cuántos puntos cayeron dentro del cuarto de círculo
    dentro_del_circulo = np.sum(distancia_cuadrada <= 1)
    
    # Aplicamos la fórmula de Monte Carlo para aproximar pi
    pi_aproximado = 4 * dentro_del_circulo / n
    
    return pi_aproximado
````

````python
print(approximate_pi(100))        # Ejemplo con 100 puntos
print(approximate_pi(1_000_000))  # Ejemplo con 1 millón de puntos
````

## Problema 5

**Descripción:**  
Creamos una función que pide al usuario un número entero positivo `n` y una base `b` entre 2 y 10.  
Luego convierte el número `n` de base 10 a la base `b` utilizando divisiones sucesivas y residuos.

Por ejemplo:

Dame un número: 17
Dame una nueva base: 3

17 base 3 = 122


**Proceso:**

$$
\begin{aligned}
17 \div 3 &= 5 \quad \text{residuo } 2 \\
5 \div 3 &= 1 \quad \text{residuo } 2 \\
1 \div 3 &= 0 \quad \text{residuo } 1 \\
\Rightarrow 122_3
\end{aligned}
$$

Mientras más divisiones se hagan, se van acumulando los residuos de **derecha a izquierda** para formar el número en la nueva base.


````python
# Creamos una función para la conversión de base
def convertir_base(n, b):
    resultado = ""  # cadena donde guardaremos el resultado en la nueva base

    # Base entre 2-10
    if b < 2 or b > 10:
        return "La base debe estar entre 2 y 10"
    
    # Mientras n sea mayor que 0, seguimos dividiendo y guardando residuos
    while n > 0:
        residuo = n % b           # obtenemos el residuo
        resultado = str(residuo) + resultado  # Agregamos al principio
        n = n // b                # Reducimos n dividiéndolo entre la base

    return resultado
````

````python
n = int(input("Dame un número: "))
b = int(input("Dame una nueva base: "))
print(f"{n} base {b} = {convertir_base(n, b)}")
````

Crea un programa que ordene ambas listas por duración del CETE.

## Problema 6

**Descripción:**  

Tenemos dos listas: una contiene la duración de varios CETES (como cadenas con días, por ejemplo `"28D"`), y la otra contiene su precio correspondiente.  
Nuestro objetivo es **ordenar ambas listas** según la duración de los CETES **de menor a mayor**, para que los precios queden correctamente alineados con la duración ordenada.

Para lograrlo:  
- 1. Extraeremos los valores numéricos de las duraciones (por ejemplo, de `"28D"` sacamos `28`).  
- 2. Emparejaremos cada duración con su precio.  
- 3. Ordenaremos las tuplas por duración.  
- 4. Finalmente, separaremos las listas ordenadas.

````python
# Definimos las listas
mty = ["28D", "256D", "56D", "364D", "84D"]
price = [9.98, 9.47, 9.85, 9.32, 9.74]

# Convertimos las duraciones a enteros quitando la "D" al final
durations = [int(x[:-1]) for x in mty]

# Usamos zip para emparejar duración con precio
cetes = list(zip(durations, mty, price))

# Ordenamos por la duración (primer elemento de cada tupla)
cetes_sorted = sorted(cetes)

# Separamos las listas ya ordenadas
sorted_mty = [m for _, m, _ in cetes_sorted]
sorted_price = [p for _, _, p in cetes_sorted]

# Mostramos resultados
print("Duraciones ordenadas:", sorted_mty)
print("Precios ordenados:", sorted_price)
````
