**1.-** Escribe un código en Python que inicialice dos vectores 𝑎 y 𝑏 de 5 elementos, con valores enteros aleatorios entre 1 y 10. Luego, realiza las siguientes operaciones:

- a) Suma de a y b
- b) Producto punto de a y b
- c) Normalizar cada vector

```python
import numpy as np

# Inicializamos vectores aleatorios
# Usamos un generador con semilla para que los resultados sean reproducibles
rng = np.random.default_rng(50)  
a = rng.integers(1, 11, size=5)  # enteros en [1, 10], 5 elementos
b = rng.integers(1, 11, size=5)

# A) Sumamos elemento a elemento
suma = a + b  # vector con la suma de cada posición

# B) Producto punto
# Equivalente: np.dot(a, b) o a.dot(b)
producto_punto = a @ b

# C) Normalizamos (L2)

def normalizar(v: np.ndarray) -> np.ndarray:
    norma = np.linalg.norm(v)  # ||v||_2
    return v / norma if norma != 0 else v  # EVITAR norma 0

a_norm = normalizar(a)
b_norm = normalizar(b)

# Resultados
np.set_printoptions(precision=4, suppress=True)  # formato bonito
print("Vector a:", a)
print("Vector b:", b)
print("Suma a + b:", suma)
print("Producto punto a · b:", producto_punto)
print("a normalizado:", a_norm)
print("b normalizado:", b_norm)
```

**2.-** En un dataset de su preferencia de 5 acciones y 300 observaciones, realiza el Análisis de Componentes Principales (PCA) para reducirlo a 2 componentes principales.

Realiza los pasos de PCA vistos en clase:

1. Centrando los datos.
2. Calculando la matriz de covarianza.
3. Obteniendo los eigenvalores y eigenvectores.
4. Seleccionando las dos componentes principales.
5. Genera el gráfico.
6. Dar análisis del gráfico generado.

**(Utilizar API YF)**

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definimos 5 acciones de ejemplo
tickers = ["KO", "TSLA", "GS", "AMZN", "META"]

# ---- Descarga robusta de precios ----
# auto_adjust=True -> precios 'Close' ya ajustados por dividendos/splits
raw = yf.download(tickers, period="430d", auto_adjust=True, progress=False)

# Si vienen columnas multinivel (OHLCV x ticker), tomamos el nivel 'Close'
if isinstance(raw.columns, pd.MultiIndex):
    prices = raw['Close']  # columnas quedan como los tickers
else:
    # Un solo nivel: quedarnos con Close; si es Serie, convertir a DataFrame
    prices = raw[['Close']] if 'Close' in raw.columns else raw.to_frame(name='Close')

# Asegurarnos de tener observaciones completas para todos los tickers
prices = prices.dropna(how='any')

# Nos quedamos con las últimas 300 observaciones (≈ 300 días hábiles)
prices = prices.tail(300)

# 1) Rendimientos log y centrado 
returns = np.log(prices / prices.shift(1)).dropna()
X_centered = returns - returns.mean()

# 2) Matriz de covarianza
cov_matrix = np.cov(X_centered.T)

# 3) Autovalores y autovectores
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

# 4) Ordenar y tomar 2 componentes
idx = np.argsort(eig_vals)[::-1]
eig_vals = eig_vals[idx]
eig_vecs = eig_vecs[:, idx]

W = eig_vecs[:, :2]               # (5x2)
X_pca = X_centered @ W            # proyección a 2D

# 5) Gráfico
loadings = pd.DataFrame(W, index=prices.columns, columns=["PC1", "PC2"])
print(loadings)

# Graficar biplot simple
plt.figure(figsize=(8,6))
plt.scatter(X_pca.iloc[:,0], X_pca.iloc[:,1], alpha=0.5, label="Observaciones")

for i, ticker in enumerate(prices.columns):
    plt.arrow(0, 0, loadings.PC1[i], loadings.PC2[i],
              color='red', alpha=0.8, head_width=0.01)
    plt.text(loadings.PC1[i]*1.1, loadings.PC2[i]*1.1, ticker, color='red')

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Biplot PCA (observaciones y loadings de acciones)")
plt.grid(True)
plt.legend()
plt.show()

# 6) Análisis de varianza explicada
var_explained = eig_vals / eig_vals.sum()
print("Varianza explicada por PC1:", round(var_explained[0]*100, 2), "%")
print("Varianza explicada por PC2:", round(var_explained[1]*100, 2), "%")
print("Varianza acumulada (PC1 + PC2):", round(var_explained[:2].sum()*100, 2), "%")
```

**6 ) Interpretación** 

- La nube de puntos concentrada cerca del origen indica que la mayoría de las variaciones son pequeñas respecto a las medias.

- Los puntos más alejados del centro son días con movimientos más extremos en combinación de varias acciones.

- La dispersión a lo largo del eje PC1 es mayor que en PC2, lo que confirma que PC1 concentra más información.

- Sobre las acciones: 

TSLA: aparece con un vector largo, separado de los demás, lo que significa que su comportamiento tiene un componente propio muy fuerte y diferente del resto.

KO: está más cerca del origen, lo que indica que contribuye menos a la varianza explicada; su rendimiento es más estable, menos volátil.

AMZN, META y GS: apuntan en direcciones similares, indicando que sus rendimientos tienden a estar más correlacionados entre sí.

**3.-** Simula 1,000 lanzamientos de un dado de 6 caras en Python y calcula la probabilidad de que la suma de dos lanzamientos consecutivos sea igual a 7. Guarda los resultados de cada lanzamiento en una lista y realiza el cálculo.

```python
import numpy as np

# Simulamos 1000 lanzamientos
# Usamos números enteros aleatorios del 1 al 6 (dado de 6 caras)

# Guardamos los resultados en una lista
np.random.seed(42)  # semilla para reproducibilidad
lanzamientos = np.random.randint(1, 7, size=1000).tolist()

# Calculamos la probabilidad de que dos lanzamientos consecutivos sumen 7
contador_eventos = 0  # contamos las veces que la suma es 7
total_pares = len(lanzamientos) - 1  # número de pares consecutivos posibles

for i in range(total_pares):
    if lanzamientos[i] + lanzamientos[i + 1] == 7:
        contador_eventos += 1

# Probabilidad 
probabilidad = contador_eventos / total_pares

# Mostramos resultados 
print("Primeros 20 lanzamientos:", lanzamientos[:20])
print(f"Número de pares consecutivos con suma 7: {contador_eventos}")
print(f"Total de pares consecutivos: {total_pares}")
print(f"Probabilidad estimada: {probabilidad:.4f}")
```

**4.-** Supón que tienes dos eventos A y B, donde:

$$P(A) = 0.6$$
$$P(B) = 0.5$$

La probabilidad de que ocurra A y B juntos es $P(A \cap  B) = 0.3$

1. Calcular la probalidad condicional de $P(A|B)$


2. Programa en Python una simulación de 1,000 casos en los que se verifican las probabilidades dadas, y calcula la probabilidad condicional a partir de la simulación para confirmar los resultados teóricos.



```python
import numpy as np

# Definimos las probabilidades dadas
p_A = 0.6
p_B = 0.5
p_AyB = 0.3
n = 1000  # número de simulaciones

# Inicializamos listas para guardar ocurrencias
A = np.zeros(n, dtype=bool)
B = np.zeros(n, dtype=bool)

np.random.seed(50)  # Damos reproducibilidad

for i in range(n):
    # Determinamos si ocurre B
    if np.random.rand() < p_B:
        B[i] = True
        # Si ocurre B, probabilidad condicional de que ocurra A es P(A|B) = 0.6
        if np.random.rand() < (p_AyB / p_B):
            A[i] = True
    else:
        # Si no ocurre B, A puede ocurrir pero no en la intersección
        # Ajustamos para que P(A) global sea 0.6
        if np.random.rand() < ((p_A - p_AyB) / (1 - p_B)):
            A[i] = True

# Estimación simulada de P(A|B)
p_A_dado_B_sim = A[B].mean()

# Resultados
print(f"P(A|B) teórica = {p_AyB / p_B:.4f}")
print(f"P(A|B) simulada = {p_A_dado_B_sim:.4f}")
```
