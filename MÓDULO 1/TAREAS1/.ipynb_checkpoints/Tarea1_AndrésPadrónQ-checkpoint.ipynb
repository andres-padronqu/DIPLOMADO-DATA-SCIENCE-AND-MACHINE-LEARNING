1. Tenemos dos listas de números enteros:

$$a = [a_0,\,a_1,\,a_2,\,...,\,a_n]\qquad\mbox{y}\qquad b = [b_0,\,b_1,\,b_2,\,...,\,b_m],$$

que cumplen $\ a_i < a_{i+1}\ $  y  $\ b_i < b_{i+1}$, es decir, las listas están ordenadas en forma ascendente. 

Crear un programa que cree una tercera lista ordenada ascendentemente que contenga todos los elementos de $a$ y de $b$ sin utilizar ningún método de ordenamiento preprogramado.

````python
#Definir las listas 
a = [1, 2, 3]
b = [-1, 3, 4, 5, 6]
````

````python
i = 0  # contador
j = 0  # contador
res = []  # lista resultado

# Comenzamos a recorrer ambas listas mientras queden elementos en ambas
while i < len(a) and j < len(b):
    if a[i] < b[j]:
        # Si el elemento de 'a' es menor, lo agregamos a 'res' y avanzamos en 'a'
        res.append(a[i])
        i += 1
    else:
        # Si el elemento de 'b' es menor o igual, lo agregamos a 'res' y avanzamos en 'b'
        res.append(b[j])
        j += 1

# Pero hay que tener cuidado, ya que pueden quedar elementos.
# Si terminamos de recorrer 'b' pero aún quedan elementos en 'a', los agregamos
while i < len(a):
    res.append(a[i])
    i += 1

# Si terminamos de recorrer 'a' pero aún quedan elementos en 'b', los agregamos
while j < len(b):
    res.append(b[j])
    j += 1

print(res)
````

2. Escribe un programa para adivinar un número que se encuentra en el intervalo $[1,\,100]$:  
Considerando un número objetivo `target_num`, el programa le deberá pedir al usuario ingresar un número `input_num`. Si `input_num` > `target_num`, el programa deberá informarle al usuario que su número está por encima del número objetivo; de otra manera, si `input_num` < `target_num`, el programa deberá informarle al usuario que su número se encuentra por debajo del número objetivo. El programa se termina una vez que el usuario adivine el número objetivo, i.e., una vez que `input_num == target_num`.

````python
# Usamos la sugerencia dada
# Importamos la librería random para generar un número aleatorio
import random

# Generamos el número objetivo entre 1 y 100
target_num = random.randint(1, 100)

print("Número aleatorio:", target_num)

while True:
    # Pedimos al usuario un número y lo convertimos a entero
    input_num = int(input("Adivina un número entre 1 y 100: "))
    
    # Si el número ingresado es mayor al objetivo
    if input_num > target_num:
        print("Tu número es mayor al número objetivo.")
    
    # Si el número ingresado es menor al objetivo
    elif input_num < target_num:
        print("Tu número es menor al número objetivo.")
    
    # Si el usuario adivinó correctamente
    else:
        print("¡Correcto! Adivinaste el número.")
        break  # Terminamos el ciclo
````

3. Modifica el programa anterior para que el usuario tenga a lo más 5 oportunidades de adivinar. En cada intento fallido se debe imprimir el número de intentos restantes y calcular una ganacia por adivinar en el próximo intento. La fórmula para calcular la ganacia es:

$$G = \frac{\$1000}{2^{n-1}}$$

donde $n$ es el número de intento. Además permita la opción de reinicar el juego si el usuario así lo desea.


Ejemplo:


```
PROGRAMA PARA ADIVINAR UN NUMERO ENTRE 1 Y 100

Adivina el número que estoy pensando, tienes 5 intentos.
Intento 1: 50
El número secreto es menor. Intento 2: 25
El número secreto es menor. Intento 3: 12
El número secreto es mayor. Intento 4: 18

¡Felicidades! Has ganado $125

Jugar de nuevo (s/n)? n

FIN DEL JUEGO


````python
# Importamos la librería sugerida
import random

print("PROGRAMA PARA ADIVINAR UN NÚMERO ENTRE 1 Y 100")

# Iniciamos el juego
seguir_jugando = "s"

while seguir_jugando == "s":
    target_num = random.randint(1, 100)  # Número a adivinar
    intentos_max = 5
    intento = 1
    adivinado = False  # Indicador si acertó

    print("\nAdivina el número que estoy pensando, tienes 5 intentos.")

    while intento <= intentos_max:
        input_num = int(input("Intento " + str(intento) + ": "))

        if input_num == target_num:
            ganancia = 1000 / (2 ** (intento - 1)) # Formula
            print("¡Felicidades! Has ganado $" + str(int(ganancia)))
            adivinado = True
            intento = intentos_max + 1  
        else:
            if input_num > target_num:
                print("El número es menor.")
            else:
                print("El número es mayor.")
            if intento < intentos_max:
                print("Intento " + str(intento + 1) + ":")
            intento = intento + 1

    # Si no adivinó en los intentos
    if adivinado == False:
        print("No lograste adivinar. El número era " + str(target_num) + ".")

    # Preguntar si quiere volver a jugar
    seguir_jugando = input("\n¿Jugar de nuevo (s/n)? ")
    if seguir_jugando != "s":
        print("\nFIN DEL JUEGO")
````

4. Tenemos en un diccionario una agenda de números telefónicos:

```
name_num = {"Arturo" : 123456, "Ana" : 234567, "Jorge" : 495872, "Eduardo" : 645322}

```

Crea un programa que pida al usuario un nombre y en caso de que se encuentre registrado devuelva el número telefónico. 

Si no está registrado, que pregunte si desea registrarlo y si es así, que pida su número (un entero de 6 dígitos) y lo agregue a la agenda.

````python
name_num = { "Arturo": 123456, "Ana": 234567, "Jorge": 495872, "Eduardo": 645322}
````

````python
# Pedimos al usuario que escriba un nombre
nombre = input("Escribe el nombre que quieres buscar en la agenda: ")

# Verificamos si el nombre ya está en la agenda
if nombre in name_num:
    print("El número de", nombre, "es:", name_num[nombre])
else:
    print(nombre, "no está en la agenda.")
    # Preguntamos si desea registrarlo
    respuesta = input("¿Deseas registrarlo? (s/n): ")
    
    if respuesta == "s":
        # Pedimos el número de 6 dígitos
        nuevo_numero = int(input("Escribe su número de 6 dígitos: "))
        
        # Lo agregamos al diccionario
        name_num[nombre] = nuevo_numero
        
        print("Contacto agregado. El número de", nombre, "es:", name_num[nombre])
    else:
        print("No se agregó ningún contacto.")
````

5. Crear un programa para transformar divisas (MXN, EUR, USD, CAD, PEN) sabiendo que:

| Divisa extrangera | Ttipo de cambio (MXN) |
|:-----------------:|:---------------------:|
| USD  | 20.26  |
| EUR | 21.16 |
| PEN | 5.39 |
| CAD | 15.71 |


Una posible solución es construir un diccionario con todos los tipos de cambio posibles suponiendo que no hay arbitrage. Esto simplifica algunos pasos.

Ejemplo 1:

```
PROGRAMA PARA CAMBIAR DE DIVISA

Monto: 250

Divisa local: MXN

Divisa foránea: USD

250 MXN = 12.34 USD
```

$$$$

Ejemplo 2:

```
PROGRAMA PARA CAMBIAR DE DIVISA

Monto: 5

Divisa local: PEN

Divisa foránea: USD

5 PEN = 1.33 USD
```

$$$$

Ejemplo 3:

```
PROGRAMA PARA CAMBIAR DE DIVISA

Monto: 24

Divisa local: PEN

Divisa foránea: COP

No se puede realizar la conversión porque la divisa COP es desconocida.
```


````python
# Diccionario con el valor en pesos mexicanos (MXN) de cada divisa
# Todas expresadas como: 1 unidad de divisa extranjera = X MXN
tipos_cambio = {"USD": 20.26, "EUR": 21.16, "PEN": 5.39, "CAD": 15.71,"MXN": 1.0}  # Se incluye MXN para facilitar conversiones entre monedas


print("PROGRAMA PARA CAMBIAR DE DIVISA\n")

# Pedimos los datos al usuario
monto = float(input("Monto: "))
divisa_local = input("Divisa local: ").upper()
divisa_objetivo = input("Divisa foránea: ").upper()

# Verificamos si ambas divisas están en el diccionario
if divisa_local in tipos_cambio and divisa_objetivo in tipos_cambio:
    # Convertimos primero a pesos mexicanos
    monto_en_mxn = monto * tipos_cambio[divisa_local]
    
    # Luego convertimos a la divisa objetivo
    monto_final = monto_en_mxn / tipos_cambio[divisa_objetivo]

    # Mostramos el resultado redondeado a 2 decimales
    print(f"\n{monto} {divisa_local} = {round(monto_final, 2)} {divisa_objetivo}")

# Si alguna divisa es desconocida
else:
    if divisa_local not in tipos_cambio:
        print(f"\nNo se puede realizar la conversión porque la divisa {divisa_local} es desconocida.")
    if divisa_objetivo not in tipos_cambio:
        print(f"\nNo se puede realizar la conversión porque la divisa {divisa_objetivo} es desconocida.")
````

[0.5 pts extra en calif final] Supongamos que en el problema anterior no se conoce la divisa CCY. Permitir al usuario agregar información del tipo de cambio de CCY a MXN y realizar la conversión deseada.

Ejemplo:

```
PROGRAMA PARA CAMBIAR DE DIVISA

Monto: 24

Divisa local: PEN

Divisa foránea: COP

No hay info de la divisa COP. Desea agregarga (Y/N) ? Y

Tipo de cambio COP/MXN: 0.0052

24 PEN = 24976.92 COP
```

````python
tipos_cambio = {"USD": 20.26, "EUR": 21.16, "PEN": 5.39, "CAD": 15.71,"MXN": 1.0}  # Se incluye MXN para facilitar conversiones entre monedas


print("PROGRAMA PARA CAMBIAR DE DIVISA\n")


# Entrada de datos
monto = float(input("Monto: "))
divisa_local = input("Divisa local: ").upper()
divisa_objetivo = input("Divisa foránea: ").upper()

# Verificamos si las divisas existen
# Si no existe una, preguntamos si quiere agregarla
if divisa_local not in tipos_cambio:
    print(f"No hay info de la divisa {divisa_local}.")
    agregar = input("¿Desea agregarla (Y/N)? ").upper()
    if agregar == "Y":
        tasa_local = float(input(f"Tipo de cambio {divisa_local}/MXN: "))
        tipos_cambio[divisa_local] = tasa_local

if divisa_objetivo not in tipos_cambio:
    print(f"No hay info de la divisa {divisa_objetivo}.")
    agregar = input("¿Desea agregarla (Y/N)? ").upper()
    if agregar == "Y":
        tasa_obj = float(input(f"Tipo de cambio {divisa_objetivo}/MXN: "))
        tipos_cambio[divisa_objetivo] = tasa_obj

# Verificamos de nuevo si ya podemos convertir
if divisa_local in tipos_cambio and divisa_objetivo in tipos_cambio:
    # Paso 1: convertir a MXN
    monto_mxn = monto * tipos_cambio[divisa_local]
    
    # Paso 2: convertir de MXN a divisa objetivo
    monto_final = monto_mxn / tipos_cambio[divisa_objetivo]

    print(f"\n{monto} {divisa_local} = {round(monto_final, 2)} {divisa_objetivo}")
else:
    print("\nNo se puede realizar la conversión por falta de datos.")
````
