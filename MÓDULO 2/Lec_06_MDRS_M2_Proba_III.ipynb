# Distribución Normal

La **Distribución Normal** es una de las distribuciones de probabilidad más importantes en estadística y probabilidad. Es una distribución continua y simétrica, también conocida como la distribución en forma de campana, debido a su característica forma. La distribución normal es fundamental en el Teorema del Límite Central y es ampliamente utilizada para modelar fenómenos naturales y sociales.

### Definición Formal

Una variable aleatoria $X$ sigue una **Distribución Normal** con media $\mu$ y desviación estándar $\sigma$ si su función de densidad de probabilidad (PDF) está dada por:
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

- $\mu$ es la **media** de la distribución, el valor promedio alrededor del cual se distribuyen los datos.
- $\sigma$ es la **desviación estándar**, que mide la dispersión de los datos alrededor de la media.

### Propiedades de la Distribución Normal

1. Es simétrica alrededor de su media $\mu$.
2. Aproximadamente el 68% de los datos se encuentran dentro de 1 desviación estándar de la media, el 95% dentro de 2 desviaciones estándar y el 99.7% dentro de 3 desviaciones estándar.
3. La media, mediana y moda coinciden en el valor $\mu$.

### Ejemplo Práctico en Python

Supongamos que queremos modelar la altura de una población con una media $\mu = 170$ cm y una desviación estándar $\sigma = 10$ cm. Este ejemplo muestra cómo:
1. Generar datos de una distribución normal.
2. Visualizar la distribución con histogramas y la curva de densidad.



```python
# Parámetros de la distribución normal
mu = 170    # media
sigma = 10  # desviación estándar

# Número de simulaciones
n_simulaciones = 1_000

# Generar muestras de una distribución normal
muestras = np.random.normal(mu, sigma, n_simulaciones)

# Graficar el histograma de las muestras
plt.figure(figsize=(12, 6))
plt.hist(muestras, bins= 40, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Graficar la curva de densidad de la distribución normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
plt.plot(x, norm.pdf(x, mu, sigma), color='red', lw=2, label=f'Distribución Normal ($\mu={mu}$, $\sigma={sigma}$)')

# Personalización del gráfico
plt.title('Distribución Normal - Altura de una Población')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

# Distribución Normal Multivariada

La **Distribución Normal Multivariada** es una extensión de la distribución normal a múltiples variables aleatorias. Es utilizada para modelar vectores de variables aleatorias que pueden estar correlacionadas entre sí. En lugar de una media y varianza, la distribución multivariada se caracteriza por un vector de medias y una matriz de covarianza.

### Definición Formal

Sea $\mathbf{X} = (X_1, X_2, \dots, X_n)$ un vector aleatorio que sigue una distribución normal multivariada con:
- $\mathbf{\mu}$, un vector de medias $\mu_i$ para cada variable aleatoria $X_i$.
- $\mathbf{\Sigma}$, una matriz de covarianza de $n \times n$ que contiene las covarianzas entre pares de variables aleatorias.

La función de densidad de probabilidad de la distribución normal multivariada es:
$$
f(\mathbf{X}) = \frac{1}{\sqrt{(2 \pi)^n |\mathbf{\Sigma}|}} \exp \left( -\frac{1}{2} (\mathbf{X} - \mathbf{\mu})^T \mathbf{\Sigma}^{-1} (\mathbf{X} - \mathbf{\mu}) \right)
$$
donde:
- $|\mathbf{\Sigma}|$ es el determinante de la matriz de covarianza.
- $\mathbf{\Sigma}^{-1}$ es la inversa de la matriz de covarianza.
  
Esta distribución es útil para modelar variables que tienen relaciones lineales y dependencias entre ellas.

### Ejemplo Visual en Python

Supongamos que tenemos dos variables aleatorias correlacionadas, como la altura y el peso de una población. Vamos a simular datos con una media de 170 cm y 65 kg para altura y peso, respectivamente, con una cierta correlación entre ambas variables.


```python
# Parámetros de la distribución normal multivariada
mu = [170, 65]  # Medias de altura y peso
cov = [[100, 80], [80, 100]] # Matriz de covarianza

# Generar datos de una distribución normal multivariada
n_simulaciones = 1000
muestras = np.random.multivariate_normal(mu, cov, n_simulaciones)

# Separar las muestras en variables individuales
altura = muestras[:, 0]
peso = muestras[:, 1]

# Crear una gráfica de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(altura, peso, alpha=0.5, color='skyblue', edgecolor='black', label='Muestras simuladas')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Distribución Normal Multivariada: Altura vs Peso')
plt.legend()
plt.grid(True)
plt.show()

# Crear un gráfico de contorno para la densidad
x, y = np.mgrid[140:200:0.5, 40:90:0.5]
pos = np.dstack((x, y))
rv = multivariate_normal(mu, cov)

plt.figure(figsize=(10, 6))
plt.contourf(x, y, rv.pdf(pos), cmap="coolwarm", levels=20)
plt.colorbar(label='Densidad')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Distribución de Densidad de una Normal Multivariada')
plt.grid(True)
plt.show()
```

# Distribución t de Student

La **Distribución t de Student** es una distribución de probabilidad continua que es particularmente útil para estimar medias de poblaciones cuando el tamaño de muestra es pequeño y/o la varianza de la población es desconocida. Se asemeja a la distribución normal, pero tiene colas más gruesas, lo que refleja mayor incertidumbre para tamaños de muestra pequeños.

### Definición Formal

La Distribución t de Student depende de un parámetro llamado **grados de libertad** ($\nu$), que afecta la forma de la distribución:
- A medida que el número de grados de libertad aumenta, la distribución t se aproxima a la **distribución normal estándar**.
- Para pocos grados de libertad, la distribución t tiene colas más gruesas, lo que refleja mayor variabilidad en muestras pequeñas.

La función de densidad de probabilidad de la distribución t de Student es:
$$
f(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \, \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$
donde:
- $\Gamma$ es la función gamma,
- $\nu$ es el número de **grados de libertad**.

### Propiedades de la Distribución t de Student

1. **Simetría**: La distribución es simétrica alrededor de 0, al igual que la distribución normal.
2. **Varianza**: La varianza es ligeramente mayor que 1 (la varianza de la normal estándar) y depende de los grados de libertad:
   - Para $\nu > 1$, la varianza es $\frac{\nu}{\nu - 2}$.
3. **Aplicación**: La distribución t es especialmente útil para realizar pruebas de hipótesis y construir intervalos de confianza cuando el tamaño de la muestra es pequeño y la varianza es desconocida.

### Ejemplo Visual en Python

Supongamos que queremos comparar la distribución t de Student con la distribución normal para diferentes grados de libertad. Esto nos permitirá visualizar cómo se comporta la distribución t a medida que aumentan los grados de libertad.


```python
# Valores de grados de libertad para comparar
grados_de_libertad = [1, 2, 5, 10, 30]

# Valores de t
x = np.linspace(-5, 5, 1000)

# Graficar la distribución normal estándar para referencia
plt.plot(x, norm.pdf(x), 'k--', lw=2, label='Normal Estándar')

# Graficar la distribución t de Student para diferentes grados de libertad
for df in grados_de_libertad:
    plt.plot(x, t.pdf(x, df), lw=2, label=f't-Student (df={df})')

# Personalización del gráfico
plt.title('Distribución t de Student vs. Distribución Normal')
plt.xlabel('Valor de t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

# Distribución Gamma

La **Distribución Gamma** es una distribución de probabilidad continua que se utiliza para modelar tiempos de espera o eventos en series continuas. Es comúnmente aplicada en modelos que describen el tiempo hasta la ocurrencia de eventos en procesos de Poisson, como la duración de vida de productos o el tiempo entre llegadas de clientes.

### Definición Formal

La distribución gamma tiene dos parámetros:
- $\alpha$ (forma): También llamado **shape**, determina la forma de la distribución.
- $\beta$ (escala): También llamado **scale**, afecta la dispersión de la distribución.

La función de densidad de probabilidad (PDF) de la distribución gamma es:
$$
f(x; \alpha, \beta) = \frac{x^{\alpha - 1} e^{-x / \beta}}{\beta^\alpha \, \Gamma(\alpha)}
$$
donde:
- $x > 0$,
- $\Gamma(\alpha)$ es la función gamma, una generalización del factorial.

### Propiedades de la Distribución Gamma

1. **Media**: La media de una distribución gamma es $E[X] = \alpha \cdot \beta$.
2. **Varianza**: La varianza es $\text{Var}(X) = \alpha \cdot \beta^2$.
3. **Aplicación**: Es útil para modelar eventos continuos y no negativos en el tiempo, como la duración de vida de productos o el tiempo entre llegadas de eventos.

### Ejemplo Visual en Python

Para visualizar la distribución gamma, generaremos gráficos para diferentes valores del parámetro de forma $\alpha$, manteniendo fijo el parámetro de escala $\beta$. Esto nos permitirá observar cómo cambia la forma de la distribución gamma según el parámetro $\alpha$.


```python
# Valores de alfa (forma) para comparar
formas = 2

# Parámetro de escala beta
beta = [1, 2, 5, 9]

# Valores de x
x = np.linspace(0, 20, 1000)

# Graficar la distribución gamma para diferentes valores de alfa
plt.figure(figsize=(12, 6))
for a in beta:
    plt.plot(x, gamma.pdf(x, a, scale=a), lw=2, label=f'Gamma (α={a}, β={a})')

# Personalización del gráfico
plt.title('Distribución Gamma para diferentes valores de α')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

#### Ejemplo: Vida Útil de una Batería de Teléfono
Supongamos que:

La vida media esperada de una batería es de 6 horas.
Sin embargo, debido a variaciones en la fabricación y en el uso, la duración puede variar.
Queremos modelar este tiempo de vida con una distribución gamma para entender la probabilidad de que una batería dure un cierto número de horas.

```python
# Parámetros de la distribución gamma
alpha = 2   # parámetro de forma
beta = 3    # parámetro de escala

# Valores de x (tiempo de vida en horas)
x = np.linspace(0, 20, 1000)

# Calcular la PDF de la distribución gamma
pdf = gamma.pdf(x, alpha, scale=beta)

# Graficar la distribución gamma
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, lw=2, label=f'Gamma (α={alpha}, β={beta})', color='blue')
plt.fill_between(x, pdf, where=(x < 4), color="red", alpha=0.3, label="Duración < 4 horas")
plt.fill_between(x, pdf, where=(x > 8), color="green", alpha=0.3, label="Duración > 8 horas")

# Personalización del gráfico
plt.title('Distribución Gamma: Vida Útil de una Batería de Teléfono')
plt.xlabel('Tiempo de Vida (horas)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Cálculos de probabilidad
prob_duracion_menor_4 = gamma.cdf(4, alpha, scale=beta)
prob_duracion_mayor_8 = 1 - gamma.cdf(8, alpha, scale=beta)

print(f"Probabilidad de que la batería dure menos de 4 horas: {prob_duracion_menor_4:.2%}")
print(f"Probabilidad de que la batería dure más de 8 horas: {prob_duracion_mayor_8:.2%}")
```

# Distribución Beta

La **Distribución Beta** es una distribución continua que toma valores en el intervalo [0, 1]. Es particularmente útil para modelar probabilidades o proporciones, como la tasa de éxito de un experimento o la fracción de tiempo que un sistema está en funcionamiento. Esta distribución es controlada por dos parámetros que determinan su forma, lo que la hace extremadamente flexible para modelar distintos tipos de datos.

### Definición Formal

La distribución beta tiene dos parámetros:
- $\alpha$ (forma 1): Controla la distribución de los valores hacia el extremo inferior (cerca de 0).
- $\beta$ (forma 2): Controla la distribución de los valores hacia el extremo superior (cerca de 1).

La función de densidad de probabilidad (PDF) de la distribución beta es:
$$
f(x; \alpha, \beta) = \frac{x^{\alpha - 1} (1 - x)^{\beta - 1}}{B(\alpha, \beta)}
$$
donde:
- $x \in [0, 1]$,
- $B(\alpha, \beta)$ es la función beta, que actúa como constante de normalización.

### Propiedades de la Distribución Beta

1. **Media**: La media de una distribución beta es $E[X] = \frac{\alpha}{\alpha + \beta}$.
2. **Varianza**: La varianza es $\text{Var}(X) = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}$.
3. **Aplicación**: Es útil para modelar probabilidades, proporciones y otras variables que están limitadas al intervalo [0, 1].

# Ejemplo Aplicado de la Distribución Beta: Tasa de Conversión en una Campaña Publicitaria

La **Distribución Beta** es útil para modelar proporciones o tasas de éxito en un intervalo de [0, 1]. Aquí aplicaremos la distribución beta para estimar la **tasa de conversión** en una campaña publicitaria en línea.

### Contexto del Ejemplo

Supongamos que una empresa está evaluando la tasa de éxito de una campaña en línea. 
- Durante la campaña, 100 personas hicieron clic en el anuncio y, de esas, **20 realizaron una compra**, resultando en una tasa de éxito observada del 20%.
- Usaremos la distribución beta para modelar esta tasa de conversión basada en los datos observados.

### Parámetros de la Distribución Beta

La distribución beta tiene dos parámetros:
- **$\alpha = \text{número de éxitos} + 1$**
- **$\beta = \text{número de fracasos} + 1$**

En este caso:
- Número de éxitos = 20, por lo que $\alpha = 20 + 1 = 21$.
- Número de fracasos = 80, por lo que $\beta = 80 + 1 = 81$.

### Visualización de la Distribución Beta

El siguiente código en Python genera la visualización de la distribución beta para estos parámetros, mostrando la probabilidad de distintas tasas de conversión y un intervalo de confianza del 95%.

```python
# Parámetros basados en los datos observados
alpha = 21  # éxitos + 1
beta_param = 81  # fracasos + 1

# Valores de x (tasa de conversión)
x = np.linspace(0, 1, 1000)

# Calcular la PDF de la distribución beta
pdf = beta.pdf(x, alpha, beta_param)

# Graficar la distribución beta
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, lw=2, color='blue', label=f'Beta (α={alpha}, β={beta_param})')
plt.fill_between(x, pdf, alpha=0.2, color='green')
plt.title('Distribución Beta: Tasa de Conversión en una Campaña Publicitaria')
plt.xlabel('Tasa de Conversión')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Calcular el intervalo de confianza (85%)
lower_bound = beta.ppf(0.075, alpha, beta_param)
upper_bound = beta.ppf(0.925, alpha, beta_param)

print(f"Intervalo de confianza del 85% para la tasa de conversión: {lower_bound:.2%} - {upper_bound:.2%}")
```

# Distribución Dirichlet

La **Distribución Dirichlet** es una distribución de probabilidad continua que generaliza la distribución beta a múltiples dimensiones. Es útil para modelar **probabilidades de eventos mutuamente excluyentes** en una serie de categorías, como proporciones en temas de documentos o probabilidades de diferentes clases en modelos de clasificación. 

### Definición Formal

La distribución Dirichlet está definida por un vector de parámetros $\alpha = (\alpha_1, \alpha_2, \dots, \alpha_K)$, donde cada $\alpha_i > 0$. Cada componente $\alpha_i$ representa una "fuerza" que tira hacia una de las categorías.

La función de densidad de probabilidad de la distribución Dirichlet para $K$ categorías es:
$$
f(x_1, x_2, \dots, x_K; \alpha) = \frac{1}{B(\alpha)} \prod_{i=1}^K x_i^{\alpha_i - 1}
$$
donde $B(\alpha)$ es la función beta multivariable que actúa como constante de normalización, y $x_i$ son las probabilidades de cada categoría tales que:
$$
\sum_{i=1}^K x_i = 1 \quad \text{y} \quad x_i \geq 0
$$

### Propiedades de la Distribución Dirichlet

1. **Media**: La media de cada componente $X_i$ es $E[X_i] = \frac{\alpha_i}{\sum_{j=1}^K \alpha_j}$.
2. **Aplicaciones**: Se usa en problemas de aprendizaje de máquina y estadística bayesiana, como en la asignación de temas en modelos de tópicos (ej. Latent Dirichlet Allocation) y en la distribución de probabilidades sobre categorías mutuamente excluyentes.

# Ejemplo Aplicado de la Distribución Dirichlet

La **Distribución Dirichlet** es útil para modelar probabilidades en categorías mutuamente excluyentes y es ampliamente utilizada en aprendizaje automático y estadística. En este ejemplo, exploraremos la distribución de probabilidades en tres categorías, como si estuviéramos modelando la distribución de temas en un conjunto de documentos.

### Parámetros de la Distribución Dirichlet

Para este ejemplo, utilizaremos:
- $\alpha = [2, 5, 3]$, donde cada valor representa el peso o "fuerza" que arrastra la probabilidad hacia una de las tres categorías:
  - $\alpha_1 = 2$, $\alpha_2 = 5$, y $\alpha_3 = 3$.

Estos valores indican que es más probable observar combinaciones de probabilidades que estén sesgadas hacia la segunda categoría.

### Visualización de la Distribución Dirichlet

El siguiente código en Python genera una visualización de la distribución Dirichlet proyectada en un triángulo equilátero, donde cada punto representa una combinación de probabilidades entre las tres categorías.

```python
# Función para simular y graficar
def graficar_dirichlet(alpha, n=2000):
    data = np.random.dirichlet(alpha, size=n)
    
    # Convertir a DataFrame
    df = pd.DataFrame(data, columns=['Acciones', 'Bonos', 'BienesRaices'])

    # Crear figura ternaria
    fig, tax = ternary.figure(scale=1.0)
    tax.boundary(linewidth=2.0)
    tax.gridlines(multiple=0.1, color="gray", linewidth=0.5)

    # Convertir a tuplas (A, B, C)
    puntos = list(df.itertuples(index=False, name=None))
    tax.scatter(puntos, alpha=0.5, s=5, color='steelblue')

    # Etiquetas
    tax.left_axis_label("Bonos", fontsize=12)
    tax.right_axis_label("Bienes Raíces", fontsize=12)
    tax.bottom_axis_label("Acciones", fontsize=12)

    tax.set_title(f"Distribución Dirichlet: α = {alpha}", fontsize=14)
    tax.ticks(axis='lbr', multiple=0.2, linewidth=1)
    tax.clear_matplotlib_ticks()
    plt.tight_layout()
    plt.show()

# Diferentes valores de α
conjuntos_alpha = [
    [1, 1, 1],       # Uniforme
    [2, 2, 2],       # Moderadamente balanceada
    [5, 5, 5],       # Más concentrada hacia el centro
    [0.5, 0.5, 0.5], # Concentrada en los vértices (extremos)
    [5, 1, 1],       # Favorece a "Acciones"
    [1, 5, 1],       # Favorece a "Bonos"
    [1, 1, 5],       # Favorece a "Bienes Raíces"
]

# Graficar cada configuración
for alpha in conjuntos_alpha:
    graficar_dirichlet(alpha)
```
