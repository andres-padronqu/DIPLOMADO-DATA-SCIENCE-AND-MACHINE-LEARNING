## Combiacion de DataFrames

Pandas ofece herramientas para combinar dataframes

- pd.merge(): combinamos datframes basandonos en valores comunes en una o mas colunas (llaves). Existen diferentes joins (inner, left, right, outer)

- pd.concat(): se utiliza para apilar dataframes uno encima del otro (por filas) o uno a lado del otro (columnas)

```python
# usamos otro ejemplo

dict_alumnos = {
    'clave_unica':[123,321,987],
    'historia':[9,10,5.6],
    'matematicas':[8.0,7.0,9.1],
}

df_alumnos =pd.DataFrame(dict_alumnos)

dict_alumnos_2 = {
    'clave_unica':[123,321,999],
    'geografia':[10,10,10],
    'literatura':[0.0,7.0,4.6],
}
df_alumnos2 = pd.DataFrame(dict_alumnos_2)
```

```python
print(df_alumnos)
print('\n')
print(df_alumnos2)
```

```python
#left join: mantiene las filas de la tabla izquierda y agrega las columnas de la tabla derecha
# si no hay match en la llave, los valores de las nuevas columnas seran NA

pd.merge(df_alumnos, df_alumnos2, on = 'clave_unica', how ='left')
```

```python
# right join: mantiene todas las filas de la tabla de la derecha y agrefa las columnas de la tabla de la izquierda

pd.merge(df_alumnos,df_alumnos2, on = ['clave_unica'], how ='right')
```

```python
# inner join: mantiene solo las filas donde hay coincidencias en la llave para las dos tablas
pd.merge(df_alumnos,df_alumnos2, on = 'clave_unica', how = 'inner')
```

```python
# Concatenado de DataFrames

#pegamos una tabla debajo de la otra (axis = 0)
# ignore_indez = True resetea e l indice del dataframe
pd.concat([df_alumnos, df_alumnos2], axis = 0, ignore_index= True)
```

```python
# Creamos una tabla al lado de la otra (axis = 1)
# se alinea por el inide, si los indices no coinciden se introduce un NA
pd.concat([df_alumnos, df_alumnos2], axis = 1)
```

## Asignaciones de nuevas Colunas

- asignacion directa: asignamos un ecalar o una serie( de la misma longitud del DF)

- basada en condiciones

- metodo apply(): definimos con funciones lambda o definidas, aplicando la funcion a cada elemento de la serie o fila/ columna del DF


```python
#ASIGNAS UN VALOR ESCALAR A TODA UNA NUEVA COLUMNA

df['px_150'] =150

#asignar true/false de una cond

df['px_mayor_150'] = df['close'] > 150

#asginar 1/0 basado en una condicion usando .apply() con una funcion lambda
df['px_mayor_150_lambda'] = df['close'].apply(lambda x: 1 if x > 150 else 0)

#asignar con una funcion definida

def es_menor_a_20(x):
    if x >= 20:
        return 0
    else:
        return 1
df['mayor_fn'] = df['close'].apply(es_menor_a_20)

df.head()
```
