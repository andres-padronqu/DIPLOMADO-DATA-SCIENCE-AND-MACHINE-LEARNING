# **Hechos y Dimensiones en un Cubo de Datos**

Un cubo de datos se construye para analizar información desde diferentes perspectivas, organizando los datos en **hechos** y **dimensiones**. 

## **Hechos (Facts o Measures)**
Los hechos son las cantidades numéricas o métricas que se desean analizar o medir. Representan los valores que pueden ser:
- **Agregados** (sumados, promediados, etc.).
- **Operados** para obtener información valiosa.

Los hechos son la razón principal para crear un cubo de datos, ya que se busca analizar estas métricas y tomar decisiones basadas en ellas.

### Ejemplo de hechos:
- **Longitud**
- **Latitud**

## **Dimensiones (Dimensions)**
Las dimensiones son atributos o categorías utilizadas para:
- **Organizar** y desglosar los datos de los hechos.
- **Proporcionar contexto** para los hechos.
  
Estas suelen ser descriptivas y no numéricas, como categorías, fechas, ubicaciones, entre otros.

### Ejemplo de dimensiones:
- **Fecha**
- **Delito**
- **Colonia**
- **Fecha de inicio**

En resumen, los **hechos** son los datos numéricos que se desean analizar, mientras que las **dimensiones** son las categorías o atributos que permiten organizar y contextualizar esos datos.

### Aplicación
Con las dimensiones mencionadas y los hechos disponibles, es posible construir un cubo de datos para analizar patrones geoespaciales y temporales relacionados con delitos, lo que facilita la toma de decisiones basadas en datos.

# **Análisis y Procesamiento Analítico en Línea (OLAP)**

El análisis de datos a gran escala, como los **351 tipos de delitos** o las **1,639 colonias**, puede ser complicado y poco práctico. Una estrategia efectiva es priorizar los elementos más relevantes mediante un **ranking** de los **10 o 20 principales elementos** en cada categoría. Esto reduce la complejidad sin perder información valiosa.

## **¿Qué es OLAP?**
**OLAP** (Online Analytical Processing o Procesamiento Analítico en Línea) es una tecnología diseñada para realizar análisis multidimensional en grandes volúmenes de datos. Proporciona herramientas y estructuras que permiten:
- **Exploración rápida de datos** desde múltiples perspectivas.
- **Consultas complejas** de manera eficiente.
- **Identificación de patrones** y tendencias dentro de los datos.

## **Características clave de OLAP**
1. **Cubo Multidimensional**: Los datos se organizan en un modelo de cubo, donde:
   - Las **dimensiones** (categorías como fecha, delito, colonia) proporcionan contexto.
   - Los **hechos** (métricas como latitud y longitud) representan los valores analizados.

2. **Agregaciones**: OLAP permite calcular métricas como sumas, promedios, máximos, mínimos y conteos sobre los datos en función de las dimensiones seleccionadas.

3. **Flexibilidad**: Facilita la interacción con los datos mediante operaciones como:
   - **Drill-down**: Profundizar en detalles más específicos.
   - **Roll-up**: Agrupar datos en niveles más generales.
   - **Slice and Dice**: Filtrar y reorganizar datos para obtener diferentes perspectivas.

## **Ventajas de OLAP**
- **Velocidad**: Responde rápidamente a consultas sobre grandes volúmenes de datos.
- **Exploración interactiva**: Permite a los analistas explorar los datos de forma dinámica.
- **Perspectivas múltiples**: Ayuda a visualizar y entender los datos desde ángulos diversos.

## **Aplicación en nuestro análisis**
Con OLAP, podemos:
- Crear cubos multidimensionales para analizar delitos geoespaciales y temporales.
- Identificar los **delitos más comunes** o las **colonias con mayor incidencia** mediante rankings.
- Descubrir patrones estacionales, como si ciertos delitos ocurren más en fechas específicas.

La tecnología OLAP responderá rápidamente a esta consulta, permitiendo un análisis profundo sin perder tiempo en procesar manualmente grandes volúmenes de datos.

En resumen, OLAP transforma conjuntos de datos extensos en información procesable y accesible, siendo una herramienta esencial para la toma de decisiones basadas en datos.

Enseguida, vamos a calcular un **hecho** que representa el tiempo transcurrido entre la denuncia y algún evento relevante. Utilizaremos la función **timedelta64** para darle un formato adecuado.

```python
df['tiempo_en_denunciar'] = (df['fecha_inicio']-df['fecha_hecho'])/np.timedelta64(1,'D')
```

Vamos a crear dos cubos de datos: uno para analizar todos los delitos sin considerar la información geoespacial y otro para aquellos que sí involucran datos geográficos.

agregamos dimensiones

```python
# Convertir la columna 'hora_hecho' a datetime
df['hora_hecho'] = pd.to_datetime(df['hora_hecho'])
# Extraer la hora de la columna 'hora_hecho'
df['hora_hecho'] = df['hora_hecho'].dt.hour
```

```python
df['fecha'] = pd.to_datetime(df['fecha_hecho']).dt.date
```

```python
df.groupby(['fecha','hora_hecho','categoria_delito','delito','alcaldia_hecho','colonia_hecho']).agg({'fecha_inicio':'count',
                                                                                                 'tiempo_en_denunciar':'sum'})
```

```python
df.groupby(['fecha','hora_hecho','categoria_delito','delito','alcaldia_hecho','colonia_hecho']).agg({'fecha_inicio':'count',
                                                                                                 'tiempo_en_denunciar':'sum'}).to_csv('cubo_delitos_sin_geo.csv')
```

En resumen, este código agrupa los datos por múltiples columnas y luego calcula la cantidad de denuncias (fecha_inicio) y la suma del tiempo en denunciar (tiempo_en_denunciar) para cada grupo
