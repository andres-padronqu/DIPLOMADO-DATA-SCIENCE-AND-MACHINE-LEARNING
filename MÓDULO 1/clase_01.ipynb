```python

```

**Nota:** Python cuenta desde cero, por lo tanto, el primer índice de `str` es `0`

1. Calcula el valor presente de un bono cupón zero con valor nominal de $C=100$, tasa de interés pagadera anual $i=7\%$ y vencimiento a $t=10$ años. El precio del bono cupón zero está dado por

$$
    P= \frac{C}{(1+i)^t}
$$

````python
C = 100
i = 0.07 #anual
t = 10 #anos

P = round(C / (1 + i)**t, 6)

print(f"El precio del bono es ${P}")
````
