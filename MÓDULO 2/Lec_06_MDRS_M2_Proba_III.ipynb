# Distribucin Normal

La **Distribucin Normal** es una de las distribuciones de probabilidad ms importantes en estadstica y probabilidad. Es una distribucin continua y simtrica, tambin conocida como la distribucin en forma de campana, debido a su caracterstica forma. La distribucin normal es fundamental en el Teorema del Lmite Central y es ampliamente utilizada para modelar fenmenos naturales y sociales.

### Definicin Formal

Una variable aleatoria $X$ sigue una **Distribucin Normal** con media $\mu$ y desviacin estndar $\sigma$ si su funcin de densidad de probabilidad (PDF) est dada por:
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

- $\mu$ es la **media** de la distribucin, el valor promedio alrededor del cual se distribuyen los datos.
- $\sigma$ es la **desviacin estndar**, que mide la dispersin de los datos alrededor de la media.

### Propiedades de la Distribucin Normal

1. Es simtrica alrededor de su media $\mu$.
2. Aproximadamente el 68% de los datos se encuentran dentro de 1 desviacin estndar de la media, el 95% dentro de 2 desviaciones estndar y el 99.7% dentro de 3 desviaciones estndar.
3. La media, mediana y moda coinciden en el valor $\mu$.

### Ejemplo Prctico en Python

Supongamos que queremos modelar la altura de una poblacin con una media $\mu = 170$ cm y una desviacin estndar $\sigma = 10$ cm. Este ejemplo muestra cmo:
1. Generar datos de una distribucin normal.
2. Visualizar la distribucin con histogramas y la curva de densidad.



```python
# Parmetros de la distribucin normal
mu = 170    # media
sigma = 10  # desviacin estndar

# Nmero de simulaciones
n_simulaciones = 1_000

# Generar muestras de una distribucin normal
muestras = np.random.normal(mu, sigma, n_simulaciones)

# Graficar el histograma de las muestras
plt.figure(figsize=(12, 6))
plt.hist(muestras, bins= 40, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Graficar la curva de densidad de la distribucin normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
plt.plot(x, norm.pdf(x, mu, sigma), color='red', lw=2, label=f'Distribucin Normal ($\mu={mu}$, $\sigma={sigma}$)')

# Personalizacin del grfico
plt.title('Distribucin Normal - Altura de una Poblacin')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

# Distribucin Normal Multivariada

La **Distribucin Normal Multivariada** es una extensin de la distribucin normal a mltiples variables aleatorias. Es utilizada para modelar vectores de variables aleatorias que pueden estar correlacionadas entre s. En lugar de una media y varianza, la distribucin multivariada se caracteriza por un vector de medias y una matriz de covarianza.

### Definicin Formal

Sea $\mathbf{X} = (X_1, X_2, \dots, X_n)$ un vector aleatorio que sigue una distribucin normal multivariada con:
- $\mathbf{\mu}$, un vector de medias $\mu_i$ para cada variable aleatoria $X_i$.
- $\mathbf{\Sigma}$, una matriz de covarianza de $n \times n$ que contiene las covarianzas entre pares de variables aleatorias.

La funcin de densidad de probabilidad de la distribucin normal multivariada es:
$$
f(\mathbf{X}) = \frac{1}{\sqrt{(2 \pi)^n |\mathbf{\Sigma}|}} \exp \left( -\frac{1}{2} (\mathbf{X} - \mathbf{\mu})^T \mathbf{\Sigma}^{-1} (\mathbf{X} - \mathbf{\mu}) \right)
$$
donde:
- $|\mathbf{\Sigma}|$ es el determinante de la matriz de covarianza.
- $\mathbf{\Sigma}^{-1}$ es la inversa de la matriz de covarianza.
  
Esta distribucin es til para modelar variables que tienen relaciones lineales y dependencias entre ellas.

### Ejemplo Visual en Python

Supongamos que tenemos dos variables aleatorias correlacionadas, como la altura y el peso de una poblacin. Vamos a simular datos con una media de 170 cm y 65 kg para altura y peso, respectivamente, con una cierta correlacin entre ambas variables.


```python
# Parmetros de la distribucin normal multivariada
mu = [170, 65]  # Medias de altura y peso
cov = [[100, 80], [80, 100]] # Matriz de covarianza

# Generar datos de una distribucin normal multivariada
n_simulaciones = 1000
muestras = np.random.multivariate_normal(mu, cov, n_simulaciones)

# Separar las muestras en variables individuales
altura = muestras[:, 0]
peso = muestras[:, 1]

# Crear una grfica de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(altura, peso, alpha=0.5, color='skyblue', edgecolor='black', label='Muestras simuladas')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Distribucin Normal Multivariada: Altura vs Peso')
plt.legend()
plt.grid(True)
plt.show()

# Crear un grfico de contorno para la densidad
x, y = np.mgrid[140:200:0.5, 40:90:0.5]
pos = np.dstack((x, y))
rv = multivariate_normal(mu, cov)

plt.figure(figsize=(10, 6))
plt.contourf(x, y, rv.pdf(pos), cmap="coolwarm", levels=20)
plt.colorbar(label='Densidad')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Distribucin de Densidad de una Normal Multivariada')
plt.grid(True)
plt.show()
```

# Distribucin t de Student

La **Distribucin t de Student** es una distribucin de probabilidad continua que es particularmente til para estimar medias de poblaciones cuando el tamao de muestra es pequeo y/o la varianza de la poblacin es desconocida. Se asemeja a la distribucin normal, pero tiene colas ms gruesas, lo que refleja mayor incertidumbre para tamaos de muestra pequeos.

### Definicin Formal

La Distribucin t de Student depende de un parmetro llamado **grados de libertad** ($\nu$), que afecta la forma de la distribucin:
- A medida que el nmero de grados de libertad aumenta, la distribucin t se aproxima a la **distribucin normal estndar**.
- Para pocos grados de libertad, la distribucin t tiene colas ms gruesas, lo que refleja mayor variabilidad en muestras pequeas.

La funcin de densidad de probabilidad de la distribucin t de Student es:
$$
f(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \, \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$
donde:
- $\Gamma$ es la funcin gamma,
- $\nu$ es el nmero de **grados de libertad**.

### Propiedades de la Distribucin t de Student

1. **Simetra**: La distribucin es simtrica alrededor de 0, al igual que la distribucin normal.
2. **Varianza**: La varianza es ligeramente mayor que 1 (la varianza de la normal estndar) y depende de los grados de libertad:
   - Para $\nu > 1$, la varianza es $\frac{\nu}{\nu - 2}$.
3. **Aplicacin**: La distribucin t es especialmente til para realizar pruebas de hiptesis y construir intervalos de confianza cuando el tamao de la muestra es pequeo y la varianza es desconocida.

### Ejemplo Visual en Python

Supongamos que queremos comparar la distribucin t de Student con la distribucin normal para diferentes grados de libertad. Esto nos permitir visualizar cmo se comporta la distribucin t a medida que aumentan los grados de libertad.


```python
# Valores de grados de libertad para comparar
grados_de_libertad = [1, 2, 5, 10, 30]

# Valores de t
x = np.linspace(-5, 5, 1000)

# Graficar la distribucin normal estndar para referencia
plt.plot(x, norm.pdf(x), 'k--', lw=2, label='Normal Estndar')

# Graficar la distribucin t de Student para diferentes grados de libertad
for df in grados_de_libertad:
    plt.plot(x, t.pdf(x, df), lw=2, label=f't-Student (df={df})')

# Personalizacin del grfico
plt.title('Distribucin t de Student vs. Distribucin Normal')
plt.xlabel('Valor de t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

# Distribucin Gamma

La **Distribucin Gamma** es una distribucin de probabilidad continua que se utiliza para modelar tiempos de espera o eventos en series continuas. Es comnmente aplicada en modelos que describen el tiempo hasta la ocurrencia de eventos en procesos de Poisson, como la duracin de vida de productos o el tiempo entre llegadas de clientes.

### Definicin Formal

La distribucin gamma tiene dos parmetros:
- $\alpha$ (forma): Tambin llamado **shape**, determina la forma de la distribucin.
- $\beta$ (escala): Tambin llamado **scale**, afecta la dispersin de la distribucin.

La funcin de densidad de probabilidad (PDF) de la distribucin gamma es:
$$
f(x; \alpha, \beta) = \frac{x^{\alpha - 1} e^{-x / \beta}}{\beta^\alpha \, \Gamma(\alpha)}
$$
donde:
- $x > 0$,
- $\Gamma(\alpha)$ es la funcin gamma, una generalizacin del factorial.

### Propiedades de la Distribucin Gamma

1. **Media**: La media de una distribucin gamma es $E[X] = \alpha \cdot \beta$.
2. **Varianza**: La varianza es $\text{Var}(X) = \alpha \cdot \beta^2$.
3. **Aplicacin**: Es til para modelar eventos continuos y no negativos en el tiempo, como la duracin de vida de productos o el tiempo entre llegadas de eventos.

### Ejemplo Visual en Python

Para visualizar la distribucin gamma, generaremos grficos para diferentes valores del parmetro de forma $\alpha$, manteniendo fijo el parmetro de escala $\beta$. Esto nos permitir observar cmo cambia la forma de la distribucin gamma segn el parmetro $\alpha$.


```python
# Valores de alfa (forma) para comparar
formas = 2

# Parmetro de escala beta
beta = [1, 2, 5, 9]

# Valores de x
x = np.linspace(0, 20, 1000)

# Graficar la distribucin gamma para diferentes valores de alfa
plt.figure(figsize=(12, 6))
for a in beta:
    plt.plot(x, gamma.pdf(x, a, scale=a), lw=2, label=f'Gamma (={a}, ={a})')

# Personalizacin del grfico
plt.title('Distribucin Gamma para diferentes valores de ')
plt.xlabel('Valor de X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

#### Ejemplo: Vida til de una Batera de Telfono
Supongamos que:

La vida media esperada de una batera es de 6 horas.
Sin embargo, debido a variaciones en la fabricacin y en el uso, la duracin puede variar.
Queremos modelar este tiempo de vida con una distribucin gamma para entender la probabilidad de que una batera dure un cierto nmero de horas.

```python
# Parmetros de la distribucin gamma
alpha = 2   # parmetro de forma
beta = 3    # parmetro de escala

# Valores de x (tiempo de vida en horas)
x = np.linspace(0, 20, 1000)

# Calcular la PDF de la distribucin gamma
pdf = gamma.pdf(x, alpha, scale=beta)

# Graficar la distribucin gamma
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, lw=2, label=f'Gamma (={alpha}, ={beta})', color='blue')
plt.fill_between(x, pdf, where=(x < 4), color="red", alpha=0.3, label="Duracin < 4 horas")
plt.fill_between(x, pdf, where=(x > 8), color="green", alpha=0.3, label="Duracin > 8 horas")

# Personalizacin del grfico
plt.title('Distribucin Gamma: Vida til de una Batera de Telfono')
plt.xlabel('Tiempo de Vida (horas)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Clculos de probabilidad
prob_duracion_menor_4 = gamma.cdf(4, alpha, scale=beta)
prob_duracion_mayor_8 = 1 - gamma.cdf(8, alpha, scale=beta)

print(f"Probabilidad de que la batera dure menos de 4 horas: {prob_duracion_menor_4:.2%}")
print(f"Probabilidad de que la batera dure ms de 8 horas: {prob_duracion_mayor_8:.2%}")
```

# Distribucin Beta

La **Distribucin Beta** es una distribucin continua que toma valores en el intervalo [0, 1]. Es particularmente til para modelar probabilidades o proporciones, como la tasa de xito de un experimento o la fraccin de tiempo que un sistema est en funcionamiento. Esta distribucin es controlada por dos parmetros que determinan su forma, lo que la hace extremadamente flexible para modelar distintos tipos de datos.

### Definicin Formal

La distribucin beta tiene dos parmetros:
- $\alpha$ (forma 1): Controla la distribucin de los valores hacia el extremo inferior (cerca de 0).
- $\beta$ (forma 2): Controla la distribucin de los valores hacia el extremo superior (cerca de 1).

La funcin de densidad de probabilidad (PDF) de la distribucin beta es:
$$
f(x; \alpha, \beta) = \frac{x^{\alpha - 1} (1 - x)^{\beta - 1}}{B(\alpha, \beta)}
$$
donde:
- $x \in [0, 1]$,
- $B(\alpha, \beta)$ es la funcin beta, que acta como constante de normalizacin.

### Propiedades de la Distribucin Beta

1. **Media**: La media de una distribucin beta es $E[X] = \frac{\alpha}{\alpha + \beta}$.
2. **Varianza**: La varianza es $\text{Var}(X) = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}$.
3. **Aplicacin**: Es til para modelar probabilidades, proporciones y otras variables que estn limitadas al intervalo [0, 1].

# Ejemplo Aplicado de la Distribucin Beta: Tasa de Conversin en una Campaa Publicitaria

La **Distribucin Beta** es til para modelar proporciones o tasas de xito en un intervalo de [0, 1]. Aqu aplicaremos la distribucin beta para estimar la **tasa de conversin** en una campaa publicitaria en lnea.

### Contexto del Ejemplo

Supongamos que una empresa est evaluando la tasa de xito de una campaa en lnea. 
- Durante la campaa, 100 personas hicieron clic en el anuncio y, de esas, **20 realizaron una compra**, resultando en una tasa de xito observada del 20%.
- Usaremos la distribucin beta para modelar esta tasa de conversin basada en los datos observados.

### Parmetros de la Distribucin Beta

La distribucin beta tiene dos parmetros:
- **$\alpha = \text{nmero de xitos} + 1$**
- **$\beta = \text{nmero de fracasos} + 1$**

En este caso:
- Nmero de xitos = 20, por lo que $\alpha = 20 + 1 = 21$.
- Nmero de fracasos = 80, por lo que $\beta = 80 + 1 = 81$.

### Visualizacin de la Distribucin Beta

El siguiente cdigo en Python genera la visualizacin de la distribucin beta para estos parmetros, mostrando la probabilidad de distintas tasas de conversin y un intervalo de confianza del 95%.

```python
# Parmetros basados en los datos observados
alpha = 21  # xitos + 1
beta_param = 81  # fracasos + 1

# Valores de x (tasa de conversin)
x = np.linspace(0, 1, 1000)

# Calcular la PDF de la distribucin beta
pdf = beta.pdf(x, alpha, beta_param)

# Graficar la distribucin beta
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, lw=2, color='blue', label=f'Beta (={alpha}, ={beta_param})')
plt.fill_between(x, pdf, alpha=0.2, color='green')
plt.title('Distribucin Beta: Tasa de Conversin en una Campaa Publicitaria')
plt.xlabel('Tasa de Conversin')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Calcular el intervalo de confianza (85%)
lower_bound = beta.ppf(0.075, alpha, beta_param)
upper_bound = beta.ppf(0.925, alpha, beta_param)

print(f"Intervalo de confianza del 85% para la tasa de conversin: {lower_bound:.2%} - {upper_bound:.2%}")
```

# Distribucin Dirichlet

La **Distribucin Dirichlet** es una distribucin de probabilidad continua que generaliza la distribucin beta a mltiples dimensiones. Es til para modelar **probabilidades de eventos mutuamente excluyentes** en una serie de categoras, como proporciones en temas de documentos o probabilidades de diferentes clases en modelos de clasificacin. 

### Definicin Formal

La distribucin Dirichlet est definida por un vector de parmetros $\alpha = (\alpha_1, \alpha_2, \dots, \alpha_K)$, donde cada $\alpha_i > 0$. Cada componente $\alpha_i$ representa una "fuerza" que tira hacia una de las categoras.

La funcin de densidad de probabilidad de la distribucin Dirichlet para $K$ categoras es:
$$
f(x_1, x_2, \dots, x_K; \alpha) = \frac{1}{B(\alpha)} \prod_{i=1}^K x_i^{\alpha_i - 1}
$$
donde $B(\alpha)$ es la funcin beta multivariable que acta como constante de normalizacin, y $x_i$ son las probabilidades de cada categora tales que:
$$
\sum_{i=1}^K x_i = 1 \quad \text{y} \quad x_i \geq 0
$$

### Propiedades de la Distribucin Dirichlet

1. **Media**: La media de cada componente $X_i$ es $E[X_i] = \frac{\alpha_i}{\sum_{j=1}^K \alpha_j}$.
2. **Aplicaciones**: Se usa en problemas de aprendizaje de mquina y estadstica bayesiana, como en la asignacin de temas en modelos de tpicos (ej. Latent Dirichlet Allocation) y en la distribucin de probabilidades sobre categoras mutuamente excluyentes.

# Ejemplo Aplicado de la Distribucin Dirichlet

La **Distribucin Dirichlet** es til para modelar probabilidades en categoras mutuamente excluyentes y es ampliamente utilizada en aprendizaje automtico y estadstica. En este ejemplo, exploraremos la distribucin de probabilidades en tres categoras, como si estuviramos modelando la distribucin de temas en un conjunto de documentos.

### Parmetros de la Distribucin Dirichlet

Para este ejemplo, utilizaremos:
- $\alpha = [2, 5, 3]$, donde cada valor representa el peso o "fuerza" que arrastra la probabilidad hacia una de las tres categoras:
  - $\alpha_1 = 2$, $\alpha_2 = 5$, y $\alpha_3 = 3$.

Estos valores indican que es ms probable observar combinaciones de probabilidades que estn sesgadas hacia la segunda categora.

### Visualizacin de la Distribucin Dirichlet

El siguiente cdigo en Python genera una visualizacin de la distribucin Dirichlet proyectada en un tringulo equiltero, donde cada punto representa una combinacin de probabilidades entre las tres categoras.

```python
# Funcin para simular y graficar
def graficar_dirichlet(alpha, n=2000):
    data = np.random.dirichlet(alpha, size=n)
    
    # Convertir a DataFrame
    df = pd.DataFrame(data, columns=['Acciones', 'Bonos', 'BienesRaices'])

    # Crear figura ternaria
    fig, tax = ternary.figure(scale=1.0)
    tax.boundary(linewidth=2.0)
    tax.gridlines(multiple=0.1, color="gray", linewidth=0.5)

    # Convertir a tuplas (A, B, C)
    puntos = list(df.itertuples(index=False, name=None))
    tax.scatter(puntos, alpha=0.5, s=5, color='steelblue')

    # Etiquetas
    tax.left_axis_label("Bonos", fontsize=12)
    tax.right_axis_label("Bienes Races", fontsize=12)
    tax.bottom_axis_label("Acciones", fontsize=12)

    tax.set_title(f"Distribucin Dirichlet:  = {alpha}", fontsize=14)
    tax.ticks(axis='lbr', multiple=0.2, linewidth=1)
    tax.clear_matplotlib_ticks()
    plt.tight_layout()
    plt.show()

# Diferentes valores de 
conjuntos_alpha = [
    [1, 1, 1],       # Uniforme
    [2, 2, 2],       # Moderadamente balanceada
    [5, 5, 5],       # Ms concentrada hacia el centro
    [0.5, 0.5, 0.5], # Concentrada en los vrtices (extremos)
    [5, 1, 1],       # Favorece a "Acciones"
    [1, 5, 1],       # Favorece a "Bonos"
    [1, 1, 5],       # Favorece a "Bienes Races"
]

# Graficar cada configuracin
for alpha in conjuntos_alpha:
    graficar_dirichlet(alpha)
```
